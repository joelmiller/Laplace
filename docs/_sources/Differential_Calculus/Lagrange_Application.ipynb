# Applications of Lagrange Multipliers: The Bradley-Terry Model

Consider a competition in which we want to rank teams or individuals based on partial information.  For example, we may want to rank all tennis players in the world, or we want to rank all teams in some league.  We often find ourselves in a situation where the teams/individuals have played each other a different number of times.  

We make a simplifying assumption that it makes sense to rank in a linear order.  That is, we assume that if team $i$ will usually beat team $j$ and team $j$ will usually beat team $k$, then team $i$ will usually beat $k$.  

We will assume that each team $i$ has a skill level $x_i$.  The probability that $i$ defeats $j$ is given by 

$$
P_{ij} = \frac{x_i}{x_i+x_j}
$$

```{prf:exercise} Absence of draws
Show that $P_{ij} + P{ji} = 1$.
```

```{prf:exercise} Consistency under re-scaling of skill

Show that if we multiply every $x_i$ by the same constant $c$, then each probability $P_{ij}$ is unchanged.
```

We can assume that $\sum x_i = 1$.


We assume that we have observed some competitions, and $W_{ij}$ denotes the number of times that team $i$ has defeated team $j$ in those competitions.

```{prf:example} A 3-team competition
Assume that we have three teams.  Team $1$ has played team $2$ $N_{12}=N_{21}=5$ times, winning $W_{12}=4$ times and losing $W_{21}=N_{12}-W_{12}=1$ time.  Team $2$ has played team $N_{23}=3$ times times and won $W_{23}=2$ times losing $W_{32}=1$ time.  Team $3$ has played team $N_{13}=1$ time, winning the competition ($P_{31}=1$).

This is demonstrated in the table below:

| |1|2|3|
|---|---|---|--:|
|1 | 0 | 4 | 0|
|2 | 1 | 0 | 2|
|3 | 1 | 1 | 0|

Assuming that each competition is an independent event, the probability of this outcome is

$$
P[\text{table}] = \binom{5}{4}P_{12}^4P_{21}^1 \binom{3}{2} P_{23}^2P_{32}^1 \binom{1}{1}  P_{31}
$$ 
with

$$
P_{ij} = \frac{x_i}{x_i+x_j}
$$
So

\begin{align*}
P[\text{table}] &= \binom{5}{4}\left(\frac{x_1}{x_1+x_2}\right)^4\left(\frac{x_2}{x_1+x_2}\right)^1 \binom{3}{2} \left(\frac{x_2}{x_2+x_3}\right)^2 \left(\frac{x_3}{x_2+x_1}\right)^1 \binom{1}{1}  \left(\frac{x_3}{x_3+x_1}\right)^1\\
&= \prod_{i\neq j} \binom{N_{i,j}}{W_{i,j}} \frac{x_i^{W_{i,j}} x_j^{W_{j,i}}}{(x_i+x_j)^{N_{i,j}}}\\
&= \left(\prod_i x_i^{W_i}\right) \left(\prod_{i\neq j} \binom{N_{i,j}}{W_{i,j}} (x_i+x_j)^{-N_{i,j}}\right)
\end{align*}
Our goal is to find the values of $x_1$, $x_2$, and $x_3$ for which $P[\text{table}]$ is largest subject to the constraint that $x_1+x_2+x_3-1=0$ (and all $x_i$ are non-negative).

To make it easier to work with, we will instad try to maximize 

\begin{align*}
\ln(P[\text{table}])
 &= \sum_i \ln(x_i^{W_i} )  + \sum_{i\neq j} \ln\binom{N_{i,j}}{W_{i,j}} + \sum_{i\neq j} \ln (x_i+x_j)^{N_{i,j}}\\
 &= \sum_i W_i \ln(x_i) + \sum_{i\neq j}\ln\binom{N_{i,j}}{W_{i,j}} + \sum_{i\neq j} -N_{i,j}\ln (x_i+x_j)
\end{align*}

To set up Lagrante Multipliers, we find that

\begin{align*}
\frac{\partial}{\partial x_i} \ln(P[\text{table}]) &= W_i \frac{1}{x_i} + \sum_{j \neq i} -N_{i,j} \frac{1}{x_i+x_j}
\end{align*}
and

$$
\frac{\partial}{\partial x_i} \sum x_j - 1 = 1
$$
So Lagrange multipliers state that for each $i$

$$
W_i \frac{1}{x_i} - \sum_{j \neq i} N_{i,j} \frac{1}{x_i+x_j} = \lambda
$$
with the constraint that $\sum x_j = 1$.

We multiply by $x_i$ to get 

$$
W_i - x_i\sum_{j \neq i} N_{i,j} \frac{1}{x_i+x_j}=  \lambda x_i
$$
Summing this, using $\sum x_i=1$, we get 

$$
\sum W_i - \sum_i \left(x_i\sum_{j \neq i}  \frac{N_{i,j}}{x_i+x_j}\right) = \lambda
$$
We can observe that $\sum_i W_i$ is the total number of wins, equal to $N=\frac{1}{2}\sum_{i}\sum_{j} N_{i,j}$.  We also observe that $x_i\sum_{j \neq i} \frac{N_{i,j}}{x_i+x_j}$ is the expected number of wins of $i$, and so summing it over $i$ also gives the expected number of wins in the entire table, also equal to $N=\frac{1}{2}\sum_i\sum_j N_{i,j}$.  So $\lambda$ must be $N-N=0$.

For each $i$, the system becomes

$$
0 = W_i - x_i\sum_{j \neq i}  \frac{N_{i,j}}{x_i+x_j}
$$
We can rewrite this as

$$
x_i = \frac{W_i}{\sum_{j \neq i} \frac{N_{i,j}}{x_i+x_j}}
$$

If appropriate conditions hold, we can solve this iteratively by assuming that initially all $x_j$ are equal, plugging the vector of values into the right hand side and finding the new vector.  If this process converges, then it gives us a solution to the equation.  

There are other methods that can be used to guarantee that this process converges.  (for example, if the vector $\vec{x}^{(n)}$ remains bounded there must be a limit).
```
