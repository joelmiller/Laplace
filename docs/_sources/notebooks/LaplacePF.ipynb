## Rational Functions

````{prf:definition} Rational Function
:label: def_rational_fxn

A **rational function** is a function that can be written in the form 

$$
f(x) = \frac{p(x)}{q(x)}
$$
where both $p(x)$ and $q(x)$ are polynomials.
````

We can always assume that the leading coefficient of $q(x)$ is $1$, because if it is not, we can divide both $p(x)$ and $q(x)$ by that coefficient to make it $1$.

We will often need to find $f(t)$ when $\tilde{f}(s) = \frac{p(s)}{q(s)}$ where $p(s)$ and $q(s)$ are both polynomials and the highest degree of $p(s)$ is less than the highest degree of $q(s)$.


## Classifying the factorizations of $q(s)$

Consider 

$$
\tilde{f}(s) = \frac{p(s)}{q(s)}
$$

where the highest degree of $p(s)$ is $0$ or $1$ and the highest degree of $q(s)$ is $2$ and the leading coefficient of $q(s)$ is $1$.  

Our goal will be to find $f(t)$.  Typically under these assumptions, $\tilde{f}(s)$ will not appear in our Laplace Transform Table.  We will find a way to write $\tilde{f}(s)$ as the sum of two functions that do appear in the table.  Our first step is to find two candidate functions and the next step is to determine the appropriate coefficients.

There are three possibilities for $q(s)$:
- $q(s) = (s-r_1)(s-r_2)$ with $r_1 \neq r_2$: there are two real unequal roots to the equation $q(s)=0$.
- $q(s) = (s-r)^2$: there is a single repeated real root to the equation $q(s) = 0$.
- $q(s) = (s-a)^2 + \omega^2$: there are no real roots to the equation $q(s)=0$.

We will use completing the square or direct factorization to determine which of these three cases applies.  Once we have classified $q(s)$ based on this, we will be able to set up Partial Fractions to find $f(t) = \mathcal{L}^{-1}[p(s)/q(s)](t)$.






### Completing the square
An important step in our process will be to take a quadratic polynomial and use "completing the square" to change its structure.  The goal is to first convert the quadratic formula into the form $(s+c_1)^2 + c_2$.  From there we will be able to convert it to one of the three forms identified above.  Rather than trying to memorize a formula for this we will try to learn a process, starting with an example.

Before you go further, please remind yourself of a common error: $(a+b)^2 \neq a^2 + b^2$.  Instead $(a+b)^2 = a^2 + 2ab + b^2$.  The $2ab$ term will play an important role in what follows.


````{prf:example} Completing the square with $q(s) = s^2 + 6s + 13$
:label: example:ssqp6sp13

We have $q(s) = s^2 + 6s+13$, and we want to write it in the form $(s+c_1)^2 + c_2$.  We make the observation that $(s+c_1)^2 = s^2 + 2c_1 s + c_1^2$.  So for the coefficient of $s$ to match, $2c_1$ must be $6$.  So $c_1=3$.  

Now we have $s^2 + 6s + 12 = (s+3)^2 + c_2$.  If we expand $(s+3)^2 = s^2 + 6s + 9$, we see that to match the constant term, we need $9+c_2=13$ so $c_2=4$.

So we now have $q(s) = (s+3)^2 + 4$.  We convert the $4$ into a square and we have

$$
q(s) = (s+3)^2 + 2^2
$$

Because this is a sum of squares rather than a difference of squares, we cannot do any more factoring (with real numbers).
````
The process we use will follow {prf:ref}`example:ssqp6sp12` 
 find $c_1$ so that the coefficient of $s$ is correct.  Then use the result to find $c_2$ so that the constant term is correct.  


````{prf:example} Completing the square with $q(s) = s^2-11s+28$
:example: ssqm11sp28

To complete the square for $q(s) = s^2-11s+28$, we follow the steps again:
\begin{align*}
(s+c_1)^2 + c_2 &= s^2-11s+28\\
s^2+2c_1 s+c_1^2 + c_2 &= s^2-11s+28
\end{align*}
So $c_1=-11/2$.
\begin{align*}
(s-11/2)^2 + c_2 &= s^2-11s+28\\
s^2-11s + \frac{121}{4} + c_2 &= s^2 - 11s+28
\end{align*}
Since $28=112/4$ we have
 $\frac{121}{4} + c_2 = \frac{112}{4}$ and $c_2 = -9/4$.  
Thus we have arrived at

$$
q(s) = (s-11/2)^2 - 9/4
$$

Unlike the previous example, this has a difference rather than a sum.  We can think of $9/4$ as $(3/2)^2$.  We can use the difference of squares factorization: $a^2-b^2 = (a-b)(a+b)$.

So 

\begin{align*}
q(s) &= \left( s- \frac{11}{2} - \frac{3}{2}\right) \left( s- \frac{11}{2} + \frac{3}{2} \right)\\
&= (s- 7)(s-4)
\end{align*}
````

````{prf:example} Completing the square for $q(s) = s^2 - 10s+25$
:label: example:ssqm10sp25

To complete the square for $q(s) = s^2 - 10s + 25$ we follow the same basic steps:
\begin{align*}
(s+c_1)^2 + c_2 &= s^2 - 10s + 25\\
s^2 + 2c_1 s + c_1^2 + c_2 &= s^2 - 10s + 25
\end{align*}
matching coefficients of $s$, we see $c_1=-5$.  This yields
\begin{align*}
s^2 - 10s + 25 + c_2 = s^2-10s+25
\end{align*}
This leads to the conclusion $c_2=0$.

So

$$
q(s) = (s-5)^2
$$
````

Using this process on any quadratic polynomial $q(s)$ with leading coefficient $1$ will lead to a factorization of $q(s)$ (if $c_2<0$) or to converting it into the form $(s-a)^2 + \omega^2$.

## Finding $\mathcal{L}^{-1}[p(s)/q(s)](t)$ if $q(s) = (s-r_1)(s-r_2)$


$$
\frac{p(s)}{q(s)} = \frac{A}{s-r_1} + \frac{B}{s-r_2}
$$

````{prf:example} 

Consider $\tilde{f}(s) = \frac{s-7}{s^2-8s+15}$.  From completing the square (not shown) or other factorization methods we get $s^2-8s+15 = (s-3)(s-5)$.  Expanding $\tilde{f}(s)$ with partial fractions:
\begin{align*}
\frac{s-7}{(s-3)(s-5)} &= \frac{A}{s-3} + \frac{B}{s-5}\\
\Rightarrow \quad s-7 &= A(s-5) + B(s-3)\\
\end{align*}
These equations are true for all values of $s$.  If we choose some convenient values, namely we can find $A$ and $B$ easily.  

For $s=3$:

$$
-4 = A(-2) \quad \Rightarrow \quad A = 2
$$
and for $s=5$

$$
-2 = B(2) \quad \Rightarrow \quad B =-1
$$

So 

$$
\tilde{f}(s) = \frac{2}{s-3} - \frac{1}{s-5}
$$
and taking the inverse transform we have

\begin{align*}
f(t) &= \mathcal{L}^{-1}\left[\frac{2}{s-3} - \frac{1}{s-5}\right](t)\\
&= 2 \mathcal{L}^{-1} \left[\frac{2}{s-3}\right](t) - \mathcal{L}^{-1} \left[\frac{1}{s-5}\right](t)\\
&= 2e^{3t} - e^{5t}
\end{align*}
````

The general approach (as above) when $q(s) = (s-r_1)(s-r_2)$ is to 
- Write 

$$
\frac{p(s)}{q(s)} = \frac{A}{s-r_1} + \frac{B}{s-r_2}
$$
- Multiply through by $q(s)$ to get 

$$p(s) = A(s-r_2)+B(s-r_1)$$
- set $s=r_1$ and solve for $A$ and set $s=r_2$ and solve for $B$.
- Invert the transform of the resulting sum.


## Finding $\mathcal{L}^{-1}[p(s)/q(s)](t)$ if $q(s) = (s-r)^2$

$$
\frac{p(s)}{q(s)} = \frac{A}{(s-r)^2}  + \frac{B}{s-r}
$$

````{prf:example}
Consider $\tilde{f}(s) = \frac{3s+14}{s^2+8s+16}$.  From completing the square (not shown) or other factorization methods we get $s^2+8s+16 = (s+4)^2$.  Expanding $\tilde{f}(s)$ with partial fractions:
\begin{align*}
\frac{3s+14}{(s+4)^2} &= \frac{A}{(s+4)^2} + \frac{B}{s+4}\\
\Rightarrow \quad 3s+14 &= A + B(s+4)\\
\end{align*}
These equations are true for all values of $s$.  The value $s=-4$ is convenient to try.  

For $s=-4$:

$$
2 = A
$$

We could try another value of $s$ (say $s=0$) which would yield $B$ with just a little Algebra.  However, it is easier at this stage to note that the coefficient of $s$ must be the same on both sides, so $B=3$.

So 

$$
\tilde{f}(s) = \frac{2}{(s+4)^2} + \frac{3}{s+4}
$$
and taking the inverse transform we have

\begin{align*}
f(t) &= \mathcal{L}^{-1}\left[\frac{2}{(s+4)^2} + \frac{3}{s+4}\right](t)\\
&= 2 \mathcal{L}^{-1} \left[\frac{2}{(s+4)^2}\right](t) + 3\mathcal{L}^{-1} \left[\frac{1}{s+4}\right](t)\\
&= 2 e^{-4t} \mathcal{L}^{-1}\left[\frac{1}{s^2}\right](t) + 3 e^{4t}\\
&= 2te^{-4t} + 3e^{4t}
\end{align*}
using the $s$-shift to get the first term in the next to last line and $t = \mathcal{L}^{-1}[1/s^2](t)$.

So
 
$$
f(t) = 2te^{-4t} + 3e^{4t}
$$
````

The general approach (as above) when $q(s) = (s-r)^2$ is to
- Write 

$$
\frac{p(s)}{q(s)} = \frac{A}{(s-r)^2} + \frac{B}{(s-r)}
$$
- Multiply through by $q(s)$ to get 

$$p(s) = A + B(s-r)$$
- Set $s=r$ to find $A$.  Match the coefficients of $s$ to find $B$ (or use another value of $s$ and solve the resulting equation).
- Invert the transform of the resulting sum (likely using the $s$-shift).

## Finding $\mathcal{L}^{-1}[p(s)/q(s)](t)$ if $q(s) = (s-a)^2+\omega^2$


$$
\frac{p(s)}{q(s)} = \frac{A(s-a)}{(s-a)^2+\omega^2}  + \frac{B\cdot \omega}{(s-a)^2+\omega^2}
$$

````{prf:example}
Consider $\tilde{f}(s) = \frac{4s+3}{s^2+4s+29}$.  From completing the square (not shown) we get $s^2+4s+29 = (s+2)^2+25$.  Expanding $\tilde{f}(s)$ with partial fractions:
\begin{align*}
\frac{4s+3}{(s+2)^2 + 25} &= \frac{A}{(s+2)^2+25} + \frac{B\cdot 5}{(s+2)^2+25}\\
\Rightarrow \quad 4s+3 &= A(s+2) + 5B\\
\end{align*}
We can look at the coefficient of $s$ and quickly conclude that

$$A=4$$

To find $B$, there are multiple options.  A good choice is to set $s=-2$ which yields $-5=5B$ so

$$B= -1$$

So 

$$
\tilde{f}(s) = \frac{4(s+2)-5}{(s+2)^2+25}
$$
and taking the inverse transform we have

\begin{align*}
f(t) &= \mathcal{L}^{-1}\left[\frac{4(s+2)-5}{(s+2)^2+25}\right](t)\\
&= 4 \mathcal{L}^{-1} \left[\frac{s+2}{(s+2)^2+25}\right](t) - \mathcal{L}^{-1} \left[\frac{5}{(s+2)^2+25}\right](t)\\
&= 4e^{-2t}\cos(5t) - e^{-2t}\cos(5t)
\end{align*}


using the $s$-shift to get the first term in the next to last line and $t = \mathcal{L}^{-1}[1/s^2](t)$.

So
 
$$
f(t) = 4e^{-2t}\cos(5t) - e^{-2t}\cos(5t)
$$
````

The general approach (as above) when $q(s) = (s-a)^2+\omega^2$ is to
- Write 

$$
\frac{p(s)}{q(s)} = \frac{A(s-a) + B\omega}{(s-a)^2+\omega^2} 
$$
- Multiply through by $q(s)$ to get 

$$p(s) = A(s-a) + B\cdot\omega$$
- Match the coefficients of $s$ to find $A$.  Then set $s=a$ to find $B$.  We can also use another value of $s$ to find $A$.
- Invert the transform of the resulting sum.


## Solving for $f(t) = \mathcal{L}^{-1}[p(s)/q(s)](t)$ when $q(s)$ is cubic or higher order

When $\tilde{f}(s) = \frac{p(s)}{q(s)}$ where $q(s)$ is cubic or higher order we again factor $q(s)$.  It is likely that we will be able to identify some of the factors of $q(s)$ directly by observing how $q(s)$ was constructed up to this point, but otherwise we may need to use more complicated methods to find the factors.

We will assume for now that we know the factorization of $q(s)$.  In general the factorization may have multiple real roots, multiple complex roots, and multiple repeated roots.  We will assume for our purposes that no complex roots are repeated, so only the real roots are repeated.

In this case, 

$$
q(s) = 
$$

To be continued.
