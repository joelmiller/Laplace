# Inverting the Laplace Transform with Partial Fractions --- quadratic denominators
## Intended Learning Outcomes
By the end of this section, you should be able to:

- Recognise when a Laplace-domain function $\tilde{f}(s) = \frac{p(s)}{q(s)}$ requires partial fraction decomposition for inversion.
- Use completing the square (or factorization) to classify a quadratic factor in the denominator into one of three standard forms:

$$
    (s-a)(s-b) \qquad (s-a)^2 \qquad (s-a) + \omega^2   
$$
- Apply partial fraction decomposition tailored to each denominator type and solve for the unknown coefficients effectively.
- Invert the resulting expressions using known transform pairs
- Extend these steps to handle functions like $\frac{e^{as}p_1(s) + p_2(s)}{q(s)}$

There are a few strategic steps in this section that you should follow in your own work to reduce algebra errors.  
- We never expand a product of polynomials when working with partial fractions.  It will be very useful to be able to identify at a glance what values of $s$ make an expression $0$.
- As we are doing the inverse transform, we take the time to write out $\mathcal{L}^{-1}[\cdots](t)$ at each step.  This prevents common errors that result from misapplying the $t$-shift or the $s$-shift.

## Introduction 

When we later apply the Laplace Transform to a differential equation for $f(t)$, we will frequently find that $\tilde{f}(s)$ is equal to $p(s)/q(s)$ where $p$ and $q$ are both polynomials (with $p$ having lower order than $q$).

Sometimes we end up with something like $\tilde{f}(s) = \frac{2}{s-1}$, which is straightforward to solve from the transform table because $\frac{1}{s-a}$ is in it.  However, often we end up with something more complicated.  For example, we might have
$\tilde{f}(s) = \frac{s+3}{s^2+3s+2}$.  We do not have something like this in our table.

We will use Partial Fractions to convert $\tilde{f}(s)$ into a sum of functions that do appear in our table.  For example, we will learn to rewrite $\tilde{f}(s) = \frac{s+3}{s^2+3s+2}$ as $\tilde{f}(s) = \frac{2}{s+1} -\frac{1}{s+2}$.  We will then use linearity to conclude $f(t) = 2\mathcal{L}^{-1}[1/(s+1)](t) - \mathcal{L}^{-1}[1/(s+2)](t)$, which reduces this to things that are in the table.  

## Rational Functions

````{prf:definition} Rational Function
:label: def_rational_fxn

A **rational function** is a function that can be written in the form 

$$
f(x) = \frac{p(x)}{q(x)}
$$
where both $p(x)$ and $q(x)$ are polynomials.
````

We can always assume that the leading coefficient of $q(x)$ is $1$.  If it is not, we can divide both $p(x)$ and $q(x)$ by that coefficient to get a new $\hat{p}(x)/\hat{q}(x)$ for which the coefficient of $\hat{q}(x)$ is $1$.  Then we use this pair as $p$ and $q$.

We will often need to find $f(t)$ when $\tilde{f}(s) = \frac{p(s)}{q(s)}$ where $p(s)$ and $q(s)$ are both polynomials and the highest degree of $p(s)$ is less than the highest degree of $q(s)$.

For our purposes, we will assume that the highest degree of $q(s)$ is always larger than the highest degree of $p(s)$.


(sec:classification)=
## Classifying the factorizations of $q(s)$

Consider 

$$
\tilde{f}(s) = \frac{p(s)}{q(s)}
$$

where the highest degree of $p(s)$ is $0$ or $1$ and the highest degree of $q(s)$ is $2$ and the leading coefficient of $q(s)$ is $1$.  

Our goal will be to find $f(t)$.  Typically under these assumptions, $\tilde{f}(s)$ will not appear in our Laplace Transform Table.  We will find a way to write $\tilde{f}(s)$ as the sum of two functions that do appear in the table.  

Our first step is to find two candidate functions to make up the sum with unknown coefficients.
The next step is to determine the appropriate coefficients.  To find those candidate functions, we first need to rewrite $q$ into one of these three possibilities:
- $q(s) = (s-a)(s-b)$ with $a \neq b$: there are two real unequal roots to the equation $q(s)=0$.
- $q(s) = (s-a)^2$: there is a single repeated real root to the equation $q(s) = 0$.
- $q(s) = (s-a)^2 + \omega^2$: there are no real roots to the equation $q(s)=0$.

If you can directly factor $q(s)$ in the first or second case, it is fine to do so.  A method that will work for all three cases is completing the square.  Once we have rewritten $q(s)$ in one of the three forms above, we will be able to set up Partial Fractions to find $f(t) = \mathcal{L}^{-1}[p(s)/q(s)](t)$.

### Completing the square
An important step in our process will be to take a quadratic polynomial and use "completing the square" to change its structure.  The goal is to first convert the quadratic formula into the form $(s+c_1)^2 + c_2$.  From there we will be able to convert it to one of the three forms identified above, depending on whether $c_2<0$, $c_2=0$, or $c_2>0$.  Rather than trying to memorize a formula for completing the square we will learn a process, starting with an example.

Before you go further, please remind yourself of a common error: $(x+y)^2 \neq x^2 + y^2$.  Instead $(x+y)^2 = x^2 + 2xy + y^2$.  The $2xy$ term plays an important role in what follows.


````{prf:example} Completing the square when $q$ does not factor: $q(s) = s^2 + 6s + 13$
:label: example:ssqp6sp13

We have $q(s) = s^2 + 6s+13$, and we want to write it in the form $(s-a)^2 + \omega$.  

We make the observation that $(s-a)^2 = s^2 -2as + a^2$.  So for the coefficient of $s$ to match, $-2a$ must be $6$.  So $a=-3$ and $(s-(-3))^2 = s^2 - (-6s) + (-3)^2 = s^2 + 6s + 9$.

We don't have $s^2+6s+9$, we have $s^2 + 6s + 13$.  There are two ways you can continue from here.  Do whichever is easiest for you to remember.

You can
- Take 9 away from the 13 to give $s^2+6s+9+4 = (s-(-3))^2+4$
- Or you can replace $s^2+6s$ by $s^2+6s+9-9$ to give $s^2+6s+13 = s^2+6s+9-9+13= (s-(-3))^2 +4$.

In either case we now have $q(s) = (s+3)^2 + 4$.  We convert the $4$ into a square and we have

$$
q(s) = (s-(-3))^2 + 2^2
$$

Because this is a sum of squares rather than a difference of squares, we cannot do any more factoring (with real numbers).
````

```{prf:algorithm} Completing the Square
The general process for completing the square for a quadratic polynomial of the form $q(s)=s^2 + cs + d$ follows {prf:ref}`example:ssqp6sp13`: 
 1. Find $a$ so that the coefficient of $s$ is $-2a=c$.  
 2. Then either
    - replace $s^2-2as$ in $q(s)$ by $s^2-2as+a^2-a^2 = (s-a)^2 -a^2$ to get $q(s) = (s-a)^2 - a^2+d$
    - or remove $a^2$ from $d$ to get $q(s)=s^2 - 2as+ a^2 + (d-a^2) = (s-a)^2 + d-a^2$.
```
It is important to remember that $(s-a)^2 = s^2-2as+a^2$ and not $s^2+a^2$.

In most examples below, we will replace $s^2-2as$ in $q(s)$ by $s^2-2as+a^2-a^2$.

````{prf:example} Completing the square when $q$ has two distinct factors: $q(s) = s^2-11s+28$
:example: ssqm11sp28

To complete the square for $q(s) = s^2-11s+28$, we follow the steps again.  We see that $-2as = -11s$, so $a=11/2$.  Then

\begin{align*}
q(s) &= s^2 - 11s + (11/2)^2 - (11/2)^2 + 28\\
     &= \left(s - \frac{11}{2}\right)^2 - \frac{121}{4} + 28\\
     &= \left(s - \frac{11}{2}\right)^2 - \frac{121}{4} + \frac{112}{4}\\
     &= \left(s - \frac{11}{2}\right)^2 - \frac{9}{4}
\end{align*}
Thus we have arrived at

$$
q(s) = (s-11/2)^2 - 9/4
$$

Unlike the previous example, this has a difference rather than a sum.  We can think of $9/4$ as $(3/2)^2$.  We can use the difference of squares factorization: $x^2-y^2 = (x-y)(x+y)$.

So 

\begin{align*}
q(s) &= \left( s- \frac{11}{2} - \frac{3}{2}\right) \left( s- \frac{11}{2} + \frac{3}{2} \right)\\
&= (s- 7)(s-4)
\end{align*}
````



````{prf:example} Completing the square when $q$ is a perfect square: $q(s) = s^2 - 10s+25$
:label: example:ssqm10sp25

To complete the square for $q(s) = s^2 - 10s + 25$ we follow the same basic steps.  We see that $-2as = -10s$. So $a=5$.  We replace $s^2 -10s$ by $s^2-10s+25-25$

\begin{align*}
q(s) &= s^2 - 10s + 25\\
     &= (s^2 - 10s +25-25) +25\\
     &= (s-5)^2 -25 + 25\\
     &= (s-5)^2
\end{align*}
````

Using this process on any quadratic polynomial $q(s)$ with leading coefficient $1$ will lead to a factorization of $q(s)$ if it has real roots or to converting it into the form $(s-a)^2 + \omega^2$.

Notice that in the case of two different real roots, we will call the final factorization $(s-a)(s-b)$, even though during the completing the square steps we used the variable $a$ to have a different meaning.  In the repeated root/complex root case, the variable $a$ in $(s-a)^2$ is the same as was used in the completing the square step.

## Finding $\mathcal{L}^{-1}[p(s)/q(s)](t)$ if $q(s) = (s-a)(s-b)$

Now that we can classify $q(s)$ based on its factorization, we will use partial fractions to break up $p(s)/q(s)$ into the sum of two things that are in the Laplace Transform Table.  If $q(s)$ has two real roots, we write:

$$
\frac{p(s)}{q(s)} = \frac{A}{s-a} + \frac{B}{s-b}
$$
Then we use Algebraic techniques to find $A$ and $B$.  We lead by example, and then summarize the steps:

````{prf:example} Finding $f(t)$ when $\tilde{f}(s) = \frac{s-7}{s^2-8s+15}$

Consider $\tilde{f}(s) = \frac{s-7}{s^2-8s+15}$.  From completing the square (not shown) or other factorization methods we get $s^2-8s+15 = (s-3)(s-5)$.  Expanding $\tilde{f}(s)$ with partial fractions:
\begin{align*}
\frac{s-7}{(s-3)(s-5)} &= \frac{A}{s-3} + \frac{B}{s-5}\qquad \text{multiply by $q(s)=(s-3)(s-5)$}\\
\Rightarrow \quad s-7 &= A(s-5) + B(s-3) \\
\end{align*}
These equations are true for all values of $s$.  If we choose some convenient values, namely $s=3$ and $s=5$ we can find $A$ and $B$ easily because these each make a term on the right hand side become $0$.  

For $s=3$

$$
-4 = A(-2) \quad \Rightarrow \quad A = 2
$$
and for $s=5$

$$
-2 = B(2) \quad \Rightarrow \quad B =-1
$$

So 

$$
\tilde{f}(s) = \frac{2}{s-3} - \frac{1}{s-5}
$$
and taking the inverse transform we have

\begin{align*}
f(t) &= \mathcal{L}^{-1}\left[\frac{2}{s-3} - \frac{1}{s-5}\right](t)\\
&= 2 \mathcal{L}^{-1} \left[\frac{2}{s-3}\right](t) - \mathcal{L}^{-1} \left[\frac{1}{s-5}\right](t)\\
&= 2e^{3t} - e^{5t}
\end{align*}
````

The general approach (as above) when $q(s) = (s-a)(s-b)$ is to 
- Write 

$$
\frac{p(s)}{q(s)} = \frac{A}{s-a} + \frac{B}{s-b}
$$
- Multiply through by $q(s)$ to get 

$$p(s) = A(s-b)+B(s-a)$$
- set $s=a$ and solve for $A$ and set $s=b$ and solve for $B$.
- Invert the transform of the resulting sum.



## Finding $\mathcal{L}^{-1}[p(s)/q(s)](t)$ if $q(s) = (s-a)^2$

$$
\frac{p(s)}{q(s)} = \frac{A}{(s-a)^2}  + \frac{B}{s-a}
$$

````{prf:example} Finding $f(t)$ when $\tilde{f}(s) = \frac{3s+14}{s^2+8s+16}$
:label: example:repeated_root

Consider $\tilde{f}(s) = \frac{3s+14}{s^2+8s+16}$.  From completing the square (not shown) or other factorization methods we get $s^2+8s+16 = (s+4)^2$.  Expanding $\tilde{f}(s)$ with partial fractions:
\begin{align*}
\frac{3s+14}{(s+4)^2} &= \frac{A}{(s+4)^2} + \frac{B}{s+4}\\
\Rightarrow \quad 3s+14 &= A + B(s+4)\\
\end{align*}
These equations are true for all values of $s$.  The value $s=-4$ is convenient to try because it makes the coefficient of $B$ become $0$. 

For $s=-4$:

$$
2 = A
$$

Next we use another value of $s$.  There is no really good second choice like there was for the previous example, but $s=0$ is often convenient.

So 

$$
14 = A + 4B= 2+4B \Rightarrow 3=B
$$

Now

$$
\tilde{f}(s) = \frac{2}{(s+4)^2} + \frac{3}{s+4}
$$
and taking the inverse transform we have

\begin{align*}
f(t) &= \mathcal{L}^{-1}\left[\frac{2}{(s+4)^2} + \frac{3}{s+4}\right](t)\\
&= 2 \mathcal{L}^{-1} \left[\frac{2}{(s+4)^2}\right](t) + 3\mathcal{L}^{-1} \left[\frac{1}{s+4}\right](t)\\
&= 2 e^{-4t} \mathcal{L}^{-1}\left[\frac{1}{s^2}\right](t) + 3 e^{-4t}\\
&= 2te^{-4t} + 3e^{-4t}
\end{align*}
We used the $s$-shift to get the first term in the next to last line and then we used $t = \mathcal{L}^{-1}[1/s^2](t)$.

So
 
$$
f(t) = 2te^{-4t} + 3e^{-4t}
$$
````

The general approach (as above) when $q(s) = (s-a)^2$ is to
- Write 

$$
\frac{p(s)}{q(s)} = \frac{A}{(s-a)^2} + \frac{B}{(s-a)}
$$
- Multiply through by $q(s)$ to get 

$$p(s) = A + B(s-a)$$
- Set $s=a$ to make the coefficient of $B$ zero and find $A$.  Then use a different value of $s$ to find an equation with $A$ and $B$ and solve for $B$.
- Invert the transform of the resulting sum (likely using the $s$-shift).

Note - in the step to find $B$, a more efficient approach is to look at the leading coefficient of $s$.  This gives $B$ directly, but the approach won't work as well when the denominator is not quadratic.

## Finding $\mathcal{L}^{-1}[p(s)/q(s)](t)$ if $q(s) = (s-a)^2+\omega^2$


$$
\frac{p(s)}{q(s)} = \frac{A(s-a)}{(s-a)^2+\omega^2}  + \frac{B\cdot \omega}{(s-a)^2+\omega^2}
$$

````{prf:example} Finding $f(t)$ when $\tilde{f}(s) = \frac{4s+3}{s^2+4s+29}$

Consider $\tilde{f}(s) = \frac{4s+3}{s^2+4s+29}$.  From completing the square (not shown) we get $s^2+4s+29 = (s+2)^2+25$.  Expanding $\tilde{f}(s)$ with partial fractions:
\begin{align*}
\frac{4s+3}{(s+2)^2 + 25} &= \frac{A}{(s+2)^2+25} + \frac{B\cdot 5}{(s+2)^2+25}\\
\Rightarrow \quad 4s+3 &= A(s+2) + 5B\\
\end{align*}

We see that setting $s=-2$ is convenient to find $B$.  We get $-5=5B$ so

$$B= -1$$

To find $A$, we choose $s=0$ to get

$$
3 = 2A + 5B = 2A-5 \Rightarrow 8=2A
$$
and so

$$A=4$$

Then

$$
\tilde{f}(s) = \frac{4(s+2)}{(s+2)^2+25} - \frac{5}{(s+2)^2+25}
$$
and taking the inverse transform we have

\begin{align*}
f(t) &= \mathcal{L}^{-1}\left[\frac{4(s+2)-5}{(s+2)^2+25}\right](t)\\
&= 4 \mathcal{L}^{-1} \left[\frac{s+2}{(s+2)^2+25}\right](t) - \mathcal{L}^{-1} \left[\frac{5}{(s+2)^2+25}\right](t)\\
&= 4e^{-2t}\cos(5t) - e^{-2t}\cos(5t)
\end{align*}


using the $s$-shift to get the first term in the next to last line and $t = \mathcal{L}^{-1}[1/s^2](t)$.

So
 
$$
f(t) = 4e^{-2t}\cos(5t) - e^{-2t}\cos(5t)
$$
````

The general approach (as above) when $q(s) = (s-a)^2+\omega^2$ is to
- Write 

$$
\frac{p(s)}{q(s)} = \frac{A(s-a) + B\omega}{(s-a)^2+\omega^2} 
$$
- Multiply through by $q(s)$ to get 

$$p(s) = A(s-a) + B\cdot\omega$$
- Set $s=a$ to find $B$.  Then use another value of $s$ to find $A$.
- Invert the transform of the resulting sum.

As in the previous example, trying to match the coefficient of $s$ is more efficient than solving the equation for the second value of $s$.  However, again it will not be as simple when our denominators are no longer quadratic.

(sec:rational_tshift)=
## Rational Functions with the $t$-shift

Sometimes we encounter functions like $\tilde{f}(s) = e^{-as}p(s)/q(s)$ where $p(s)$ and $q(s)$ are polynomials.  We will still approach this using Partial Fractions, but first we need to apply the $t$-shift.

````{prf:example} Finding $f(t)$ when $\tilde{f}(s) = \frac{e^{-3s}(3s+14)}{s^2+8s+16}$

For $\tilde{f}(s) = \frac{e^{-3s}(3s+14)}{s^2+8s+16}$, when we try to find $f(t)$, we will first use the $t$-shift:
\begin{align*}
f(t) &= \mathcal{L}^{-1}\left[\frac{e^{-3s}(3s+14)}{s^2+8s+16}\right](t)\\ 
&=  H_3(t)\mathcal{L}^{-1}\left[\frac{3s+14}{s^2+8s+16}\right](t-3)
\end{align*}

Now, if we can find $\mathcal{L}^{-1}\left[\frac{3s+14}{s^2+8s+16}\right](t)$, then we just have to substitute $t-3$ and multiply by $H_3(t)$ to get the solution to the original problem.  This is a rational function, and the techniques we learned above apply.  

We can use Partial Fractions to complete this.
In fact, in {prf:ref}`example:repeated_root` we saw this example already and we found that

$$
\mathcal{L}^{-1}\left[\frac{3s+14}{s^2+8s+16}\right](t) = 2te^{-4t} + 3e^{-4t}
$$

Thus our solution is 
\begin{align*}
f(t) &= H_3(t) \left( 2(t-3)e^{-4(t-3)} + 3 e^{-4(t-3)} \right)\\
&= H_3(t) \left(2t e^{-4(t-3)} - 3e^{-4(t-3)}\right)
\end{align*}

````

In general when $\tilde{f}(s)$ is an exponential times a rational function, we apply the $t$-shift first, and then we use Partial Fractions on the remaining part.
If $\tilde{f}(s)$ involves a sum of rational functions times exponentials, we apply this approach to each term separately.  

````{prf:example} Finding $f(t)$ when $\tilde{f}(s) =  \frac{e^{-3s}(3s+14) + s-1}{s^2+8s+16}$

For $\tilde{f}(s) = \frac{e^{-3s}(3s+14) + e^{-2s}(s-1)}{s^2+8s+16}$, when we try to find $f(t)$, we first use linearity to break it into terms with different exponents.

\begin{align*}
f(t) &= \mathcal{L}^{-1} \left[\frac{e^{-3s}(3s+14) + e^{-2s}(s-1)}{s^2+8s+16}\right](t) \\
&= \mathcal{L}^{-1} \left[\frac{e^{-3s}(3s+14)}{s^2+8s+16}\right] +\mathcal{L}^{-1} \left[\frac{e^{-2s}(s-1)}{s^2+8s+16}\right](t)\\
&=H_3(t) \mathcal{L}^{-1} \left[ \frac{3s+14}{s^2+8s+16}\right](t-3) + H_2(t) \mathcal{L}^{-1}\left[\frac{(s-1)}{s^2+8s+16}\right](t-2)
\end{align*}
We now deal with each term in the sum separately.  We have done the first one in an earlier example.  We need to find $\mathcal{L}^{-1}\left[\frac{(s-1)}{s^2+8s+16}\right](t)$.  We use partial fractions:

\begin{align*}
\frac{(s-1)}{s^2+8s+16} &= \frac{A}{(s+4)^2} + \frac{B}{s+4}\\
s-1 &= A + B(s+4)\\
\end{align*}
Taking $s=-4$ gives $A=-5$.  Taking any other value of $s$, say $s=1$, gives $A+5B=0$, from which $A=-5$ implies $B=1$.  (alternately, we could match coefficients of $s$ to find $B=1$).

So

\begin{align*}
\mathcal{L}^{-1}\left[\frac{(s-1)}{s^2+8s+16}\right](t) &= \mathcal{L}^{-1} \left[ \frac{-5}{(s+4)^2} + \frac{1}{s+4} \right](t)\\
&= -5\mathcal{L}^{-1} \left[ \frac{1}{(s+4)^2}\right](t) + \mathcal{L}^{-1}\left[\frac{1}{s+4} \right](t)\\
&= -5e^{-4t} \mathcal{L}^{-1}\left[\frac{1}{s^2}\right](t) + e^{-4t}\\
&= -5 e^{-4t} t + e^{-4t}
\end{align*}
Combining our results above,

\begin{align*}
f(t) &= H_3(t) \mathcal{L}^{-1} \left[ \frac{3s+14}{s^2+8s+16}\right](t-3) + H_2(t) \mathcal{L}^{-1}\left[\frac{(s-1)}{s^2+8s+16}\right](t-2)\\
&= H_3(t) \left(2t e^{-4(t-3)} - 3 e^{-4(t-3)}\right) + H_2(t) \left(-5(t-2)e^{-4(t-2)} + e^{-4*(t-2)} \right)
\end{align*}
````


## Self-Test
1. For each of the following functions:
   - Use completing the square to rewrite the denominator of each of the following functions into one of the forms $(s-a)(s-b)$,  $(s-a)^2$, or $(s-a)^2+\omega^2$.  Use completing the square even if you can spot a factorization some other way.
   -  Write down the appropriate partial fraction expansion (with unknown coefficients $A$ and $B$).
   - Find the coefficients.
   - Find $f(t)$.
````{tab-set}
   ```{tab-item} $\tilde{f}(s)=\frac{2s-12}{s^2-6s+13}$
      $\tilde{f}(s)=\frac{2s-12}{s^2-6s+13}$
   ```
   ```{tab-item} Denominator 
      ````{tab-set}
         ```{tab-item} Solution 
            $(s-3)^2 + 2^2$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            s^2 -6s + 13&= s^2-6s+9-9 + 13 
            &= (s-3)^2 + 4\\
            &= (s-3)^2+2^2
            \end{align*}
         ```
      ````
   ```
   ```{tab-item} Partial Fraction
      $$\frac{2s-12}{(s-3)^2 + 2^2} = \frac{A(s-3)}{(s-3)^2 + 2^2} + \frac{2B}{(s-3)^2+2^2}$$
   ```
   ```{tab-item} Coefficients
      ````{tab-set}
         ```{tab-item} Solution 
            $A=2$,  $B=-3$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            \frac{2s-12}{(s-3)^2 + 2^2} &= \frac{A(s-3)}{(s-3)^2 + 2^2} + \frac{2B}{(s-3)^2+2^2}\\
            2s-12 &= A(s-3) + 2B
            \end{align*}
            $s=3$:
            
            $$
            -6 = 2B \quad \Rightarrow B=-3
            $$
            $s=0$:

            $$-12 =-3A-6 \quad \Rightarrow A=2$$
         ```
      ````
   ```
   ```{tab-item} $f(t)$
      ````{tab-set}
         ```{tab-item} Solution 
            $f(t) = 2 e^{3t} \cos(2t) - 3 e^{3t}\sin(2t)$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            f(t) & = \mathcal{L}^{-1}\left[ \frac{2(s-3)}{(s-3)^2 + 2^2} + \frac{(-3)2}{(s-3)^2+2^2}\right](t)\\
            &= 2 \mathcal{L}^{-1} \left[\frac{s-3}{(s-3)^2 + 2^2}\right](t) -3 \mathcal{L}^{-1} \left[ \frac{2}{(s-3)^2 + 2^2}\right](t)\\
            &= 2 e^{3t} \cos(2t) - 3 e^{3t}\sin(2t)
            \end{align*}
         ```
      ````
   ```
````    
````{tab-set}
   ```{tab-item} $\tilde{f}(s)=\frac{5s+8}{s^2+2s-8}$
      $\tilde{f}(s)=\frac{5s+8}{s^2+2s-8}$
   ```
   ```{tab-item} Denominator 
      ````{tab-set}
         ```{tab-item} Solution 
            $$(s+4)(s-2)$$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            s^2+2s-8 &= s^2+2s+1-1-8\\
            &= (s-(-1))^2 -9 & \\
            &= (s+1)^2-9\\
            &= (s+1+3)(s+1-3)\\
            &= (s+4)(s-2)
            \end{align*}
         ```
      ````
   ```
   ```{tab-item} Partial Fraction
      $$\frac{A}{s+4} + \frac{B}{s-2}$$
   ```
   ```{tab-item} Coefficients
      ````{tab-set}
         ```{tab-item} Solution 
            $A=2$,  $B=3$ 
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            \frac{5s+8}{(s+4)(s-2)} &= \frac{A}{s+4} + \frac{B}{s-2}\\
            5s+8 &= A(s-2) + B(s+4)
            \end{align*}
            $s=2$:

            $$
            18 = 6B \quad \Rightarrow B=3
            $$
            $s=-4$:

            $$
            -12 = -6A \quad \Rightarrow A=2
            $$
         
         ```
      ````
   ```
   ```{tab-item} $f(t)$
      ````{tab-set}
         ```{tab-item} Solution 
            $f(t) = 2e^{-4t} + 3 e^{2t}$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            f(t) &= \mathcal{L}^{-1} \left[ \frac{2}{s+4} + \frac{3}{s-2}\right](t)\\
            &= 2 e^{-4t} + 3e^{2t}
            \end{align*}
         ```
      ````
   ```
```` 
````{tab-set}
   ```{tab-item} $\tilde{f}(s)=\frac{2}{s^2+6s+8}$
      $\tilde{f}(s)=\frac{2}{s^2+6s+8}$
   ```
   ```{tab-item} Denominator 
      ````{tab-set}
         ```{tab-item} Solution 
            $(s+4)(s+2)$
         ``` 
         ```{tab-item} Worked solution
            \begin{align*}
            s^2+6s+8 &= s^2+6s+9-9+8\\
            &= (s-(-3))^2 -1\\
            &= (s+3)^2-1 \\
            &= (s+3+1)(s+3-1)\\
            &= (s+4)(s+2)
            \end{align*}
         ```
      ````
   ```
   ```{tab-item} Partial Fraction
      $$\frac{A}{s+4} + \frac{B}{s+2}$$
   ```
   ```{tab-item} Coefficients
      ````{tab-set}
         ```{tab-item} Solution 
            $A=-1$, $B=1$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            \frac{2}{(s+4)(s+2)} &= \frac{A}{s+4} + \frac{B}{s+2}\\
            2 &= A(s+2) + B(s+4)
            \end{align*}
            $s=-2$: 

            $$
             2 = 2B \quad \Rightarrow B=1
            $$
            $s=-4$:

            $$
            -2A \quad \Rightarrow A=-1
            $$
         ```
      ````
   ```
   ```{tab-item} $f(t)$
      ````{tab-set}
         ```{tab-item} Solution 
            $f(t) =-e^{-4t} + e^{-2t} $
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            f(t) &= \mathcal{L}^{1}\left[ \frac{-1}{s+4}  + \frac{1}{s+2}\right](t)\\
             &= -e^{-4t} + e^{-2t}
            \end{align*}
         ```
      ````
   ``` 
````    
````{tab-set}
```{tab-item} $\tilde{f}(s)=\frac{3s+3}{s^2+4s+4}$
$\tilde{f}(s)=\frac{3s+3}{s^2+4s+4}$
```
   ```{tab-item} Denominator 
      ````{tab-set}
         ```{tab-item} Solution 
         (s+2)^2
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            s^2+4s+4 &= s^2+4s+4-4+4
            &= (s+2)^2 +0\\
            &= (s+2)^2
            \end{align*}
         ```
      ````
   ```
   ```{tab-item} Partial Fraction
      $$\frac{A}{(s+2)^2} + \frac{B}{s+2}$$
   ```
   ```{tab-item} Coefficients
      ````{tab-set}
         ```{tab-item} Solution 
            $A=-3$, $B=3$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            \frac{3s+3}{(s+2)^2} &= \frac{A}{(s+2)^2} + \frac{B}{s+2}\\
            3s+3 &= A + B(s+2)
            \end{align*}
            $s=-2$: 

            $$
             -3 = A \quad \Rightarrow A=-3
            $$
            $s=0$:

            $$
            3 = A + 2B \quad \Rightarrow 3 = -3+2B \quad \Rightarrow B=3
            $$
         ```
      ````
   ```
   ```{tab-item} $f(t)$
      ````{tab-set}
         ```{tab-item} Solution 
            $ f(t) = -3 te^{-2t} + 3e^{2t}$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            f(t) &= \mathcal{L}^{-1} \left[\frac{-3}{(s+2)^2} + \frac{3}{s+2} \right](t)\\
            &= \mathcal{L}^{-1} \left[\frac{-3}{(s+2)^2}\right](t) + \mathcal{L}^{-1}\left[\frac{3}{s+2} \right](t)\\
            &= e^{-2t}\mathcal{L}^{-1}\left[\frac{-3}{s^2}](t) + 3 e^{-2t}\\
            &= -3 te^{-2t} + 3e^{2t}
            \end{align*}
         ```
      ````
   ```
````   
````{tab-set}
```{tab-item} $\tilde{f}(s)=\frac{-2s}{s^2+2s+10}$
$\tilde{f}(s)=\frac{-2s}{s^2+2s+10}$
```
   ```{tab-item} Denominator 
      ````{tab-set}
         ```{tab-item} Solution 
         $(s+1)^2+3^2$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            s^2+2s+10&= s^2+2s+1-1+10\\
            &= (s+1)^2 + 9\\
            &= (s+1)^2 + 3^2
            \end{align*}
         ```
      ````
   ```
   ```{tab-item} Partial Fraction
      $$\frac{A(s+1) + 3B}{(s+1)^2+3^2}$$
   ```
   ```{tab-item} Coefficients
      ````{tab-set}
         ```{tab-item} Solution 
            $A=-2$, $B=2/3$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            \frac{-2s}{(s+1)^2+3^2} &= \frac{A(s+1) + 3B}{(s+1)^2+3^2}\\
            -2s &= A(s+1) + 3B
            \end{align*}
            $s=-1$: 

            $$
             2 = 3B \quad \Rightarrow B=\frac{2}{3}
            $$
            $s=0$:

            $$
            0 = A + 2 \quad \Rightarrow A = -2
            $$
         ```
      ````
   ```
   ```{tab-item} $f(t)$
      ````{tab-set}
         ```{tab-item} Solution 
            $f(t)= -2 e^{-t}\cos(3t) + \frac{2}{3} e^{-t} \sin(3t) $
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            f(t) &= \mathcal{L}^{-1}\left[\frac{-2(s+1) + 3\frac{2}{3}}{(s+1)^2+3^2}\right](t)\\
            &= -2 e^{-t}\cos(3t) + \frac{2}{3} e^{-t} \sin(3t)
            \end{align*}
         ```
      ````
   ```
````    
````{tab-set}
```{tab-item} $\tilde{f}(s)=\frac{3s+1}{s^2-3s-28}$
$\tilde{f}(s)=\frac{3s+1}{s^2-3s-28}$
```
   ```{tab-item} Denominator 
      ````{tab-set}
         ```{tab-item} Solution 
         $(s+4)(s-7)$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            s^2-3s- 28&= s^2-3s + \left(\frac{3}{2}\right)^2 - \left(\frac{3}{2}\right)^2 - 28\\
            &= (s-3/2)^2 - \frac{9}{4} - \frac{112}{4} \\
            &=\left(s-\frac{3}{2}\right)^2 - \frac{121}{4}\\
            &= \left(s-\frac{3}{2}+\frac{11}{2}\right)\left(s-\frac{3}{2}-\frac{11}{2}\right)\\
            &= (s+4)(s-7)
            \end{align*}
         ```
      ````
   ```
   ```{tab-item} Partial Fraction
      $$\frac{A}{s+4} + \frac{B}{s-7}$$
   ```
   ```{tab-item} Coefficients
      ````{tab-set}
         ```{tab-item} Solution 
            $A=1$, $B=2$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            \frac{3s+1}{(s+4)(s-7)} &= \frac{A}{s+4} + \frac{B}{s-7}\\
            3s+1 &= A(s-7) + B(s+4)
            \end{align*}
            $s=7$: 

            $$
             22 = 11B \quad \Rightarrow B=2
            $$
            $s=-4$:

            $$
            -11 =-11A \quad \Rightarrow A=1
            $$
         ```
      ````
   ```
   ```{tab-item} $f(t)$
      ````{tab-set}
         ```{tab-item} Solution 
            $ f(t)=e^{-4t} + 2 e^{7t}$
         ```
         ```{tab-item} Worked solution
            \begin{align*}
            
            f(t) &= \mathcal{L}^{-1} \left[ \frac{1}{s+4} + \frac{2}{s-7}\right](t)\\
            &= e^{-4t} + 2 e^{7t}
            \end{align*}
         ```
      ````
   ```
````     


2. Find $f(t)$ if

   $$ \tilde{f}(s) = \frac{2s-12}{s^2-6s+13} + e^{-3s}\frac{2}{s^2+6s+8} -2 e^{-s}\frac{3s+1}{s^2-3s-28}$$

   Your answers in the earlier questions should be helpful.  If you have trouble, revisit {numref}`sec:rational_tshift`.
3.  Using techniques we will learn soon, Susan applies the Laplace Transform to $f''(t) - f(t) = 0$ with $f(0)=-1$ and $f'(0)=5$.  She finds that $\tilde{f}(s) = \frac{s-5}{s^2-1}$, but she has forgotten how to do Partial Fractions.  Find $f(t)$ for Susan.
