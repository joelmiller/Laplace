# Systematically finding the terminals

In the motivating example we used for multiple integrals, we were able to write the integral as 

$$
\iint_S f(x,y) = \int_{-1}^1 \int_{3x^2-3}^{x^2-1} f(x,y) dy \, dx
$$

It is useful to note that there are typically multiple possible ways to describe a two-dimensional region $S$.  For our purposes, it is useful to note that the integral we write down corresponds to one of these options:

$$
S = \left \{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l} 3x^2-3 \leq y \leq x^2-1\\
-1 \leq x \leq 1
\end{array}
\right. \right\}
$$
where here and for the remainder of this section, we will usually write conditions in a column to save space.
We could arrive at this description of $S$ by noting that if we pre-specify $x=x_0$, then we know that the set of all possible $y$ values is given by $ 3x_0^2-3 \leq y \leq x_0^2-1$, and further, we know that the global bounds on $x$ are $-1 \leq x \leq 1$.

For this particular shape, if we were to try to create a similar description where the bounds of $x$ depend on $y$ instead, then we would specify $-3 \leq y \leq 0$ (since those are the minimum/maximum of $y$ over $S$), but we would not be able to write bounds on $x$ in the form $x_{min}(y) \leq x \leq x_{max}(y)$, specifically because of the region $-1 < y \leq 0$ for which there are two distinct regions of $x$.  This is why $S$ needed to be divided into three parts to do the integral.


Now we will try to generalize this process to other regions.



## 2D case

Let us consider an integral over some two-dimensional region $S$.  

- If we can write $S$ as 

$$
S = \left\{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l}
y_{min}(x) \leq y \leq y_{max}(x)\\
x_{min} \leq x \leq x_{max}
\end{array} \right. \right\}
$$
where $x_{min}$ and $x_{max}$ are constants, then 

$$ 
\iint_S f(x,y) \, dS = \int_{x_{min}}^{x_{max}} \left[\int_{y_{min}(x)}^{y_{max}(x)} f(x,y) \, dy \right] \, dx
$$

- If we can write $S$ as 

$$
S = \left\{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l}
x_{min}(y) \leq x \leq x_{max}(y)\\
y_{min} \leq y \leq y_{max}
\end{array} \right. \right\}
$$
where $y_{min}$ and $y_{max}$ are constants, then 

$$ 
\iint_S f(x,y) \, dS = \int_{y_{min}}^{y_{max}} \left[\int_{x_{min}(y)}^{x_{max}(y)} f(x,y) \, dx \right] \, dy
$$

Often we can do either of the versions above.  However, sometimes only one is possible, or perhaps neither is possible.



### The steps 

**Add illustrative sketch**

We almost always start with a sketch of the region, then we choose the order of our variables.  We will assume that the bounds on the first variable depends on the second, but that the bounds of the second are global constants.  With experience, we will learn to see what order is the best from just looking at the sketch.

In the description below, we assume our first variable is $x$ and the second is $y$.  Otherwise, we follow the same steps, but with their roles swapped:
1) we write down bounds on the first variable, in terms of the second.  So assuming our first variable to be $x$, we have $x_{min}(y) \leq x \leq x_{max}(y)$.  It must be true that for each $y$ that occurs in $S$, the point $(x,y)$ is in $S$ if and only if $ x_{min}(y) \leq x \leq x_{max}(y)$.  So our motivating example at the start of this section would not work for this order.
2) Then we write down the bounds on our second variable using the minimum and maximum of that variable in the region $S$.  So assuming our second variable is $y$, we have $y_{min} \leq y \leq y_{max}$ where $y_{min}$ and $y_{max}$ are constants.  Then

$$
S= \left\{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l} 
x_{min}(y) \leq x \leq x_{max}(y)\\
y_{min} \leq y \leq y_{max}
\end{array} \right. \right\}
$$

3) If it is not possible to describe $S$ using bounds like this, we either swap the order of our variables or we have to divide $S$ into smaller regions for which this is possible.

4) If we have described $S$ using these bounds then 

$$\iint_S f(x,y) \, dS = \int_{y_{min}}^{y_{max}} \left[\int_{x_{min}(y)}^{x_{max}(y)} f(x,y) \, dx \right]\, dy$$ 


The process described above yields $dS = dx \, dy$.  If asked to find the integral with $dS = dy \, dx$, you would swap the roles of $x$ and $y$.

````{prf:example}  $S=\{ (x,y) \in \mathbb{R}^2 | x(x-1) \leq y \leq x\}$ with $dS = dy \, dx$
:label: example:2Dmultint2a

Consider the region 

$$ 
S = \{ (x,y) \in \mathbb{R}^2 | x(x-1) \leq y \leq x\} 
$$
using $dS = dy \, dx$

In this case the bounds on $y$ are already in the desired form: $x(x-1) \leq y \leq x$.  

We just need to figure out the minimum and maximum value of $x$.  This requires finding where the two curves intersect.  That is, where $x = x(x-1)$.   

One approach to solve this equation is to divide by $x$.  If we do that, we have to check what happens if the thing we divide by is $0$.  We see that $x=0$ is a solution.  So we have $x=0$ or if $x \neq 0$, we can divide and get $1 = x-1$, in which case $x=2$.  So we have two solutions $x=0$ or $x=2$.  Our bounds are $0 \leq x \leq 2$.

An alternate approach is to expand both sides and rearrange to get $x = x^2-x$, so $0 = x^2-2x = x(x-2)$.  From this it follows that $x=0$ or $x=2$.  In either case we end up with

$$
S = \left \{ (x,y) \in \mathbb{R}^2 \left | \begin{array}{l} 
x(x-1) \leq y \leq x\\
0 \leq x \leq 2
\end{array} \right. \right \}
$$
So the integral is

$$
\iint_S f(x,y) \, dS = \int_0^2 \int_{x^2-x}^x f(x,y) \, dy \, dx
$$
````
Now we consider the opposite direction:
````{prf:example}  $S=\{ (x,y) \in \mathbb{R}^2 | x(x-1) \leq y \leq x\}$ with $dS = dy \, dx$
:label: example:2Dmultint2a

Consider the region 

$$ 
S = \{ (x,y) \in \mathbb{R}^2 | x(x-1) \leq y \leq x\} 
$$
using $dS = dx \, dy$

For this, we need to figure out the boundaries on $x$ as functions of $y$.  The upper boundary of $x$ is always given by the right part of the parabola $y = x(x-1)$.  Let's rewrite the right part of the parabola $y = x(x-1)$ to give $x$ in terms of $y$.  We have

\begin{align*}
y &= x^2 - x\\
0 &= x^2 - x - y\\
\Rightarrow x &= \frac{1 \pm \sqrt{1 +4y}}{2}
\end{align*}
As we want the upper boundary of $x$, we choose the $+$ in the numerator and get

$$
x_{max}(y) = \frac{1 + \sqrt{1+4y}}{2}
$$
Now we look at the left.  We see that 
- for $y\leq 0$, the boundary is the left part of the parabola, which is $ x_{min}(y) = \frac{1-\sqrt{1+4y}}{2}$, but this only works for $y\leq 0$.
- for $y>0$, the boundary is the line $y=x$, or $x=y$.  So for $y>0$, we have $x_{min}(y) = y$.

The simplest way to handle this is to divide the region $S$ into two sub-regions, $S_1$ for which $y>0$ and $S_2$ for which $y \leq 0$.  Looking closer at each region:
- in $S_1$: $0 < y \leq 2$
- in $S_2$: $-1/4 \leq y \leq 0$.  

So we have $S = S_1 \cup S_2$ where

\begin{align*}
S_1 &= \left \{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l}
y \leq x \leq \frac{1 + \sqrt{1+4y}}{2}\\
0 \leq y \leq 2 
\end{array}
\right. \right\}\\
S_2 &= \left \{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l}
\frac{1 - \sqrt{1+4y}}{2} \leq x \leq \frac{1 + \sqrt{1+4y}}{2}\\
-\frac{1}{4} \leq y \leq 0
\end{array} \right. \right\}
\end{align*}


So our integral becomes

\begin{align*}
\iint_S f(x,y) \, dS &= \iint_{S_1} f(x,y) \, dS_1  + \iint_{S_2} f(x,y) \, dS_2\\
&= \int_0^2 \int_y^{\frac{1 + \sqrt{1+4y}}{2}} f(x,y) \, dx \, dy + \int_{1/4}^0 \int_{\frac{1 - \sqrt{1+4y}}{2}}^{\frac{1 + \sqrt{1+4y}}{2}} f(x,y) \, dx \, dy
\end{align*}
````
Typically for the example above, we would choose to do the integral with $dS = dy \, dx$ rather than $dS = dx \, dy$.


Finally, we consider an example for which it is impossible to avoid breaking the region up into parts.

````{prf:example}  $S=\{ (x,y) \in \mathbb{R}^2 | x(x-1) \leq y \leq x\}$ with $dS = dy \, dx$
:label: example:2Dmultint3

Consider the region 

$$ 1 \leq x^2 + y^2 \leq 4$$

with $dS = dx \, dy$.

SKETCH

This region is an annulus.  It is a disk of radius $2$ with the central disk of radius $1$ removed.  It has a hole in the middle.  

If we try to find bounds on $x$ in terms of $y$, there is a problem for the values of $y$ between $-1$ and $1$.  We will divide $S$ into four regions, one ($S_1$) with $-2 \leq y \leq -1$, two with $-1 < y < 1$ (using positive $x$ for $S_2$ and negative $x$ for $S_3$), and one ($S_4$) with $1 \leq y\leq 2$.

We will write the left and right hand side of each circle as $x$ as a function of $y$.
- For $x^2+y^2=4$, we have $x^2 = 4-y^2$, so $x = \pm \sqrt{4-y^2}$.
- For $x^2+y^2=1$, we have $x^2 = 1-y^2$, so $x = \pm \sqrt{1-y^2}$.

SKETCH

From this we conclude that 
\begin{align*}
S_1 &= \left \{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l}
-\sqrt{4-y^2} \leq x \leq \sqrt{4-y^2}\\
-2 \leq y \leq -1 
\end{array} \right. \right\}\\
S_2&= \left \{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l}
-\sqrt{4-y^2} \leq x \leq -\sqrt{1-y^2}\\
-1 < y< -1 
\end{array} \right. \right\}\\
S_3&= \left \{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l}
\sqrt{1-y^2} \leq x \leq \sqrt{4-y^2}\\
-1< y < -1 
\end{array} \right. \right\}\\
S_4&= \left \{ (x,y) \in \mathbb{R}^2 \left| \begin{array}{l}
-\sqrt{4-y^2} \leq x \leq \sqrt{4-y^2}\\
1\leq y \leq 2
\end{array} \right. \right\}
\end{align*}

Thus using $dS = dx \, dy$,
\begin{align*}
\iint_S f(x,y) dS = \quad &\iint_{S_1} f(x,y) d{S_1} +\iint_{S_2} f(x,y) d{S_2} +\iint_{S_3} f(x,y) d{S_3} +\iint_{S_4} f(x,y) d{S_4} \\
= \quad &\int_{-2}^{-1} \int_{-\sqrt{4-y^2}}^{\sqrt{4-y^2}} f(x,y) \, dx \, dy\\
& + \int_{-1}^{1} \int_{-\sqrt{4-y^2}}^{-\sqrt{1-y^2}} f(x,y) \, dx \, dy\\
& + \int_{-1}^{1} \int_{\sqrt{1-y^2}}^{\sqrt{4-y^2}} f(x,y) \, dx \, dy\\
& + \int_{1}^{2} \int_{-\sqrt{4-y^2}}^{\sqrt{4-y^2}} f(x,y) \, dx \, dy
\end{align*}
````

Due to symmetry, if we did $dS = dy \, dx$, we would get an equivalent result.





### 3D case

In 3D, we follow a similar process.  We choose an order to the three variables.  We find bounds on our first variable in terms of the other two.  Then we find bound on the second variable in terms of the third.  Finally we have a global bound on the third.  So, for example if we want $dV = dz \, dy \, dx$,  we look to describe $V$ in the form

$$
V = \left \{ (x,y,z) \in \mathbb{R}^3 \left| \begin{array}{c}
z_{min}(x,y) \leq z \leq z_{max}(x,y) \\
y_{min}(x) \leq y \leq y_{max}(x)\\
x_{min} \leq x \leq x_{max}  
\end{array}\right.
\right\}
$$
If this is possible, then we can write down our integral.  If not, then we must divide our region into parts or try a different variable order.

````{prf:example}  $V=\{ (x,y,z) \in \mathbb{R}^3 |  x^2 + (y-1)^2 + (z-2)^2 \leq 4^2 \}$ with $dV = dx \, dy \, dz$
:label: example:3DmultintSphere

Consider the region

$$V=\{ (x,y,z) \in \mathbb{R}^3 |  x^2 + (y-1)^2 + (z-2)^2 \leq 4^2 \}$$

with $dV = dx \, dy \, dz$.

We first find bound on $x$ in terms of $y$ and $z$.  We have

\begin{align*}
&x^2 \leq 16 - (y-1)^2 - (z-2)^2\\
&\Rightarrow -\sqrt{16 - (y-1)^2 - (z-2)^2} \leq x \leq \sqrt{16 - (y-1)^2 - (z-2)^2}
\end{align*}
So we have bounds on $x$ in terms of $y$ and $z$.  Now we move to $y$.  We have

$$
(y-1)^2 \leq 16 - x^2 - (z-2)^2 
$$
We cannot simply take square roots here because we would end up with bounds on $y$ that depends on $x$.  We want bounds for $y$ in terms of $z$ only.  On further consideration, we see that if $z$ were chosen, then we could choose any $y$ such that $(y-1)^2 \leq 16-(z-2)^2$, and then the possible $x$ would follow from our bounds above.  So we have
$
-\sqrt{16-(z-2)^2} \leq y-1 \leq \sqrt{16-(z-2)^2}
$
which becomes

$$
1- \sqrt{16-(z-2)^2} \leq y \leq 1 + \sqrt{16-(z-2)^2}
$$

Finally, we find $(z-2)^2 \leq 16$, so $-4 \leq z-2 \leq 4$ or

$$
-2 \leq z \leq 6
$$

So we finally describe $V$ as

$$
V = \left \{ (x,y,z) \in \mathbb{R}^3 \left| \begin{array}{c}
-\sqrt{16-(y-1)^2-(z-2)^2} \leq x \leq \sqrt{16-(y-1)^2-(z-2)^2}\\
1-\sqrt{16 - (z-2)^2} \leq y \leq 1 + \sqrt{16 - (z-2)^2}\\
-2 \leq z \leq 6 
\end{array}
\right. \right\}
$$
Thus 

$$\iiint_V f(x,y,z) \, dV = \int_{-2}^6 \int_{1-\sqrt{16 - (z-2)^2}}^{1+\sqrt{16 - (z-2)^2}} \int_{-\sqrt{16-(y-1)^2-(z-2)^2}}^{\sqrt{16-(y-1)^2-(z-2)^2}} f(x,y,z) \, dx \, dy \, dz
$$

````
