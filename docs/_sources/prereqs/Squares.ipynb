# Squares
Some of the most common errors that occurs when students do Algebra are related to squares.

Please ensure that you are extremely comfortable with these concepts.  In a sport the goal isn't to do a skill in isolation, but to internalize it and combine it with other skills in a competitive environment.  Similarly, here it's not enough to be able to do these correctly in isolation, but rather to internalize them so that it is automatic. 


## The square of a sum
A common mistake is to simplify $(a(x)+b(x))^2$ by assuming that the square of a sum is the same as the sum of a square: 

$$
(a(x)+b(x))^2 = a(x)^2 + b(x)^2  \quad {\color{red} \leftarrow \text{Incorrect assumption}}
$$
Please do not do this.  

To illustrate what's happening, let's assume $a$ and $b$ are positive and consider the following square with side length $a+b$.  The area of the square is $(a+b)^2$.  However, you can divide the square into 4 pieces: a square of side $a$ (and area $a^2$), a square of side $b$ (and area $b^2$), and two rectangles of sides $a$ and $b$ (with area $ab$).  

```{figure} ../images/(a+b)^2.png
---
height: 250px
name: fig:square_of_sum
---
The area of a square with side length $a+b$ is $a^2 + 2ab+b^2$.  Treating it as $a^2+b^2$ accounts for the two squares (red and blue), but not the two rectangles (grey).
```

If you treat $(a+b)^2$ as $a^2+b^2$, you are claiming that the area of those two rectangles is zero.

The drawing above is awkward if the values can be negative, so to show it in more generality, here's an Algebraic derivation:

\begin{align*}
(a+b)^2 &= (a+b)(a+b)\\
&= a(a+b) + b(a+b)\\
&= a^2 + ab + ba +b^2\\
&= a^2 + 2ab+ b^2
\end{align*}
You should memorize that $(a+b)^2 = a^2+2ab+b^2$.  This will take practice.

## The square root of a sum

Another common error occurs in square roots:

$$\sqrt{a(x)^2 + b(x)^2} = a(x)+b(x) \quad  {\color{red} \leftarrow \text{incorrect assumption}}$$


You can easily check that this assumption is incorrect.  Consider any example where $a$ and $b$ are both nonzero.  Say $a=1$, $b=1$:

$$
\sqrt{1^2+1^2} = \sqrt{1+1} = \sqrt{2}
$$
But if you make the error you get:

$$
\sqrt{1^2+1^2} {\color{red} \,=\, } 1+1 = 2
$$
So either $\sqrt{2}=2$ (in fact repeated applications of this argument would conclude $\sqrt{3}=3$, $\sqrt{4}=4$, etc.) or there is an error.  

When you encounter something like $f(x) = \sqrt{a(x)^2 + b(x)^2}$ there are only really only two ways to get rid of the square root:
- do Algebra steps with $a(x)^2+b(x)^2$ to arrive at a simplified expression that can be written as $c(x)^2$, in which case $\sqrt{a(x)^2+b(x)^2} = |c(x)|$.
- square both sides of the equation to get $f(x)^2 = a(x)^2 + b(x)^2$.



## Solving for $x$ if we know $x^2$

It often occurs that we get something of the form $x^2=K$ (here $K$ must be 0 or positive).  It is natural to assume that

$$
\text{If } x^2=K \text{ then } x = \sqrt{K}  \quad  {\color{red} \leftarrow \text{incorrect assumption}}
$$

However, here is one counter-example: if $x^2=4$, then $x$ could be $-2$ instead of $2$.  We're now at a level of mathematics where it's not enough to know some of the solutions to your equation.  You usually need to have all of them.  You must use

$$
\text{If } x^2=K \text{ then } x = \pm \sqrt{K}
$$

A similar problem can occur when we have 

$$y^2=x^2$$
Again it is natural to assume that $y = \sqrt{x^2}=x$.  There are two problems that can occur here.  Both of them are solved by taking

$$y = \pm x$$

The first problem is that if $x<0$, then $\sqrt{x^2} \neq x$.  Intead $\sqrt{x^2} = |x|$.  The second problem is that if we have $y^2 = x^2$, then $y = \pm \sqrt{x^2} = \pm |x|$.  Since $\pm |x| = \pm x$, we can simplify to $y = \pm x$.

(2.1) $\sqrt{x^2 + 9} \stackrel{?}{=} x + 3$

````{tab-set}
```{tab-item} Expression
Determine whether $\sqrt{x^2 + 9} = x + 3$ is valid for all real $x$.
```

```{tab-item} Solution
(a) Not valid for all $x$.

(b) Counterexample: $x=1$ gives  
- LHS: $\sqrt{10}$  
- RHS: $4$

(c) The square root of a sum is **not** equal to the sum of square roots.

(d) $x=0$
```

```{tab-item} Worked Solution
(a) We test whether  

$$(x^2 + 9)^{1/2} = x + 3$$

holds for all $x$.

(b) Choose $x=1$:

- LHS: $\sqrt{1^2 + 9} = \sqrt{10}$  
- RHS: $1 + 3 = 4$

Since $\sqrt{10} \neq 4$, the identity fails.

(c) The error comes from assuming  

$$\sqrt{a^2 + b^2} = a + b,$$

which is not a valid property of square roots.

(d) If $(x^2+9)^{1/2} = x+3$ then by squaring both sides

\begin{align*} 
x^2+9 &= x^2 + 6x + 9\\
0&= 6x\\
0 &=x
\end{align*}
Because squaring both sides of an equation can introduce incorrect solutions, we check that $x=0$ is a solution.  Tryin $x=0$ in the original expresion gives $3=3$.
```
````

(2.3) $\sqrt{(x+1)^2 + (x+2)^2} \stackrel{?}{=} (x+1)+(x+2)$

````{tab-set}
```{tab-item} Expression
Determine whether  
$\sqrt{(x+1)^2 + (x+2)^2} = (x+1)+(x+2)$.
```

```{tab-item} Solution
(a) Not valid for all $x$.

(b) Counterexample: $x = 0$  
- LHS: $\sqrt{5}$  
- RHS: $3$

(c) The square root cannot be distributed across addition.

(d) $x=-1$  (note $x=-2$ is not a solution)  
```

```{tab-item} Worked Solution
(a) We examine whether

$$\sqrt{(x+1)^2 + (x+2)^2} = (x+1)+(x+2)$$

is always true.

(b) Let $x=0$:

- LHS: $\sqrt{1^2 + 2^2} = \sqrt{5}$  
- RHS: $1 + 2 = 3$

Since $\sqrt{5} \neq 3$, the identity is false.

(c) This fails because  

$$\sqrt{A + B} \neq \sqrt{A} + \sqrt{B}$$

(unless one of $A$ and $B$ is zero and the other one is non-negative).

(d) We will look at three different ways to find the solution, in order of increasing sophistication and decreasing difficulty.  

- If
$\sqrt{(x+1)^2 + (x+2)^2} = (x+1)+(x+2)$ then

    \begin{align*}
    \sqrt{(x+1)^2 + (x+2)^2} &= 2x+3\\
    \Rightarrow (x+1)^2 + (x+2)^2 &= (2x+3)^2\\
    x^2+2x+1 + x^2+4x+4 &= 4x^2 + 12x + 9\\
    2x^2 + 6x + 5 &= 4x^2 + 12x + 9\\
    0 &= 2x^2 + 6x + 4\\
    0 &= x^2 + 3x+2\\
    0 &= (x+2)(x+1) 
    \end{align*}

    So $x =-2$ or $x=-1$.  However, when we squared both sides of the equation, we might have introduced some incorrect solutions.  That is, the step marked with $\Rightarrow$ is not necessarily reversible.  There may be values of $x$ that make the second equation true but not the first.  We need to check our values in the original equation.  In this case, if $x=-1$, the equation is $\sqrt{0^2 + 1^2} = 0+1$, which is correct.  However, for $x=-2$, the equation is $\sqrt{(-1)^2+0^2} = -1 + 0$ which is not true.  So $x=-2$ is not a solution.

- If we are patient and don't try to combine everything.  

    \begin{align*}
    \sqrt{(x+1)^2 + (x+2)^2} &= (x+1) + (x+2)\\
    \Rightarrow (x+1)^2 + (x+2)^2 &= (x+1)^2 + 2(x+1)(x+2) + (x+2)^2\\
    \Rightarrow 0 &= 2(x+1)(x+2)
    \end{align*}
    So $x =-2$ or $x=-1$.  The same argument as in the previous attempt excludes $x=-2$.

- Finally, if $\sqrt{a^2+b^2}=a+b$ then either ($a=0$ and $b \geq 0$) or ($a \geq 0$ and $b=0$).  That is, either $x+1=0$ and $x+2\geq 0$, which occurs for $x=-1$ or $x+1 \geq 0$ and $x+2=0$.  But $x+2=0$ means $x+1<0$, so this is not possible.  Thus $x=-1$ is the only solution. 
```
````
