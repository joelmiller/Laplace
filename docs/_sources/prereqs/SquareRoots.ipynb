# Square roots

One of the most common errors we encounter is the following FALSE assumption.

$\sqrt{a(x)^2 + b(x)^2} = a(x)+b(x)  {\color{red} \leftarrow \text{incorrect assumption}}$

This error is one of the clearest indicators of who will or will not pass MAT2VCA.  It is one thing to know that this is not true (which thankfully most students know) and another thing to not do it when under time pressure (which many students do).  Whether or not students make this error on their final exam is strongly predictive of whether they pass the subject comfortably or not.  

You can easily check that this assumption is incorrect.  Consider any example where $a$ and $b$ are both nonzero.  Say $a=1$, $b=1$:

$$
\sqrt{1^2+1^2} = \sqrt{1+1} = \sqrt{2}
$$
But if you make the error you get:

$$
\sqrt{1^2+1^2} {\color{red} = } 1+1 = 2
$$
So either there is an error or $\sqrt{2}=2$ (in fact repeated applications of this argument would conclude $\sqrt{3}=3$, $\sqrt{4}=4$, etc.).  There is an error.

When you encounter something like $f(x) = \sqrt{a(x)^2 + b(x)^2}$ there are only really only two ways to get rid of the square root:
- do Algebra steps with $a(x)^2+b(x)^2$ to arrive at a simplified expression that can be written as $c(x)^2$, in which case $\sqrt{a(x)^2+b(x)^2} = |c(x)|$.
- square both sides of the equation to get $f(x)^2 = a(x)^2 + b(x)^2$.
