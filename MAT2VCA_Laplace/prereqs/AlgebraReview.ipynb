# How to improve your prerequisite skills

If you've ever played a sport, you probably know that there are certain skills that are needed to reach the next level.  You also know that to master a skill, you need repetition.  It's not enough to watch the coach demonstrate a technique and then move on.  It's not enough to play a video game where your digital self can do it automatically.

You have to do it yourself until you master it.

Someone an athlete moves up to a higher level and discovers that they are missing skills that are expected.  To succeed, they need to do extra work to master the missing skills while simultaneously working on the new things their coach is teaching.  This is very difficult.

Many students who arrive in Vector Calculus but are missing important Algebra or single variable Calculus skills.  They need to master these while learning the new Vector Calculus material.

Hopefully, if you're in this position, you have a chance to improve your missing skills before the semester starts.

## Recipe for mastering Algebra:

Here is a somewhat exaggerated recipe for how to get your Algebra skills ready.

1. Set a pot of water on the stove on medium heat
2. Order an Algebra workbook.
3. When the water boils, place your calculator or any other device that can do Algebra for you into the pot.  
5. Add large amounts of salt.  
6. Stir vigorously until it returns to a boil
7. Boil for 10 minutes.
8. Work through the workbook in detail.  
   - Do not be afraid to seek out help from others when you encounter difficulties.
   - Place your focus on developing fluency in the skill.  You should eventually be able to look at an Algebra problem and identify what techniques will be needed.

(don't actually follow these steps.  They are meant to illustrate the fact that developing a dependence on calculators has harmed many students.)
