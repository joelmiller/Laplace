# The use of $=$ and $\Rightarrow$

Mathematical writing has grammar rules, though they are not frequently explicitly taught.  Some of the most common errors occur when people misuse either $=$ or $\Rightarrow$.

In mathematical writing, the equals sign takes a specific meaning.  Unfortunately, it is different from what the equals sign means in primary school and what it means with a calculator.  This leads to confusion.

In primary school, you would see something like $3+4=$ and you would be expected to fill in $7$.  For most calculators, the equals sign plays a similar role. You type the keys in order: $3+4=$ and the calculator says $7$.  So many students are used to the idea that the equals sign is the instruction "perform the calculation".  If we want to go back and check our own work, or we want others to be able to read our work and understand our reasoning, we should not do this.


In mathematical writing, the equals sign sits in the middle of an equation.  An equation is an assertion or claim that the two things on either side of the equals sign are in fact equal to one another.  


Consider now $\Rightarrow$.  This is used to denote that if one statement is True, then the next statement must also be True.  


Most mathematical writing of the variety that shows up in MAT2VCA should be based on a sequence of equations.  Typically we start with one equation, then we do some mathematical operation to both sides of the equation and arrive at a new equation.  

Let us consider an illustrative example showing a correct use of the two symbols and then an incorrect use.  Both demonstrate correct reasoning and get to the same answer.  However, the first uses mathematical grammar correctly.  The second is more like a string of incomplete sentences.

Here is a sample problem: *Consider $f(x) = x^2 -6x + 9$.  Find $f(3)$.*

Here are two correct ways to write the solution:
\begin{align*}
f(x) &= x^2 - 6x + 9\\
\Rightarrow f(3) &= 3^2 - 6\times 3 + 9\\
\Rightarrow f(3) &= 9 - 18 + 9\\
\Rightarrow f(3) &= 0
\end{align*}
A shortcut we use in mathematics when we are just simplifying one side of an equation is to write a sequence of equals signs, without needing the $\Rightarrow$:
\begin{align*}
f(x) &= x^2-6x+9\\
\Rightarrow f(3) &= 3^2-6\times 3 + 9\\
&= 9 - 18 + 9\\
&= 0
\end{align*}
Either of the solutions above is acceptible.  Notice that the mathematical operations happen between each line.  There is a specific calculation that is done from one equation to the next.  

Here is an incorrect way of writing the solution.  It matches exactly the sequence of steps you would use if using a (standard) calculator.  The thought process underlying these steps is correct.  However, the presentation is incorrect.  

$f(3)=3^2 = 9 - 6\times 3 = 9-18=-9+9=0$

If you look at it, you can probably determine the steps that are actually being taken.  But as written, this makes a series of claims such as

$$
3^2 = 9 - 6\times 3
$$


The source of the problems with doing a calculation like this is that it's hard to know what each equals sign really means.  In the "correct" versions higher up, I can ask you to verify the step between the second and third line and you can check exactly what calculation was done and whether it was done correctly.  In this version, if I ask you to check the fourth equals sign ($9-18=-9+9$), you can't do that in isolation, you have to go through the entire calculation to see what that step is.

This makes it difficult for a student to go back and check for accuracy.  It makes it difficult for a marker to understand the student's thought process.  More concerningly, it increases the risk of errors.  Imagine looking at medical notes and seeing something like this by a doctor calculating a child's medication dosage.

If you write down an equals sign and the two things on either side of it are not equal, you will lose points.
