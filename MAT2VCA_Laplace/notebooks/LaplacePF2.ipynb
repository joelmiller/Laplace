# Inverting the Laplace Transform with Partial Fractions --- cubic and higher order denominators

## Intended Learning Outcomes
By the end of this section, you should be able to:
- Identify the appropriate Partial Fractions expansion of $\tilde{f}(s) = p(s)/q(s)$ where $q(s)$ is cubic or higher order and the degree of $p$ is less than the degree of $q$.
-  Set up the appropriate partial fraction expansion by recognizing and matching different types of denominator factors:
    - Simple real roots: $s-a$
    - Repeated real roots: $(s-a)^n$
    - Irreducible quadratic factors: $(s-a)^2 + \omega^2$
-  Solve for the unknown partial fraction coefficients 
- Use known Laplace transform pairs to invert each term—combining results to find $f(t) = \mathcal{L}^{-1}[\tilde{f}(s)](t)$

## Introduction

When $\tilde{f}(s) = \frac{p(s)}{q(s)}$ where $q(s)$ is cubic or higher order and $p(s)$ has lower order than $q(s)$, the process is similar to the quadratic case.  We will assume that we know the factored version of $q(s)$ and build up a partial fractions expansion for $p(s)/q(s)$.  

We will consider three cases for the factors of $q(s)$:
- $(s-a)$ appears as a single unrepeated factor of $q(s)$.  In this case the Partial Fraction expansion will have a term of the form 

$$\frac{A}{s-a}$$

- $(s-a)^n$ for $n>1$ appears as a factor of $q(s)$.  In this case, the Partial Fraction expansion will have terms of the form

$$ \frac{A_1}{s-a} + \frac{A_2}{(s-a)^2} + \cdots + \frac{A_n}{(s-a)^n}$$

- $(s-a)^2 + \omega^2$ appears as a factor of $q(s)$.  In this case, the Partial Fraction expansion will have a term of the form

$$ \frac{A(s-a) + B\omega}{(s-a)^2+\omega^2}$$

We are ignoring the possibility that $(s-a)^2 + \omega^2$ is repeated.  We will address this case in an extension exercise.

Once we have built up the full Partial Fraction expansion with terms from all factors of $q(s)$, we use the same approach to find the coefficients.  Then we can invert the transform to find $f(t)$.

Notice that in all of the examples, we leave all of the expressions in factored form.  We never expand the product of polynomials.  Often we will use $A$, $B$, $C$, etc rather than $A_1$, $A_2$, $A_3$ etc.  As long as we use a different symbol for each term, it is unimportant whether we use subscripts or different letters.


## Distinct Real Roots

When all roots are distinct, we can easily find all coefficients in the partial fractions expansion.

````{prf:example} Distinct real roots
:label: example:distinct

Find $f(t)$ if $\tilde{f}(s) =  \frac{7s^2-31s+36}{(s-1)(s-2)(s-3)}$

$$
\frac{p(s)}{q(s)} = \frac{7s^2-31s+36}{(s-1)(s-2)(s-3)}
$$

We set
\begin{align*}
\frac{p(s)}{q(s)} &= \frac{A}{s-1} + \frac{B}{s-2} + \frac{C}{s-3}\\
\Rightarrow \quad 
7s^2-31s+36 &= A(s-2)(s-3) + B(s-1)(s-3) + C(s-1)(s-2)
\end{align*}
Now we set $s=1$ and get

$$
12 = A(-1)(-2) \quad \Rightarrow \quad A= 6
$$

Next we set $s=2$ and get

$$
2 = B(1)(-1)  \quad \Rightarrow \quad B = -2
$$

Finally we set $s=3$ and get

$$ 
6= C(2)(1) \quad \Rightarrow \quad C=3
$$

So 

$$
\frac{p(s)}{q(s)} = \frac{6}{s-1} - \frac{2}{s-2} + \frac{3}{s-3}
$$
and

$$
f(t) = 6e^t - 2 e^{2t} + 3e^{3t}
$$

The solution $f(t)$ is plotted below.
````



`````matlab
t = linspace(0, 1, 1001);
plot(t, 6*exp(t)-2*exp(2*t)+3*(exp(3*t)))
xlabel('$t$', 'Interpreter', 'latex')
ylabel('$f(t)$', 'Interpreter', 'latex')
`````

## Repeated roots

When we have a repeated root, the technique of guessing a value of $s$ to quickly get an equation for just a single coefficient will not be sufficient.  It is advisable to use this trick where possible to get as many coefficients as feasible.  However, we will need additional equations.  

We have several possible sources for these equations.  The most common are  
- Substitute different values of $s$ that seem convenient to get equations with more than one of the unknown coefficients.
- Match coefficients of powers of $s$ (usually it is advisable to start with the highest powers).  

Ultimately, we need as many equations as coefficients.[^lin_indep_note].

[^lin_indep_note]: Specifically we need the same number of *linearly independent* equations as we have coefficients.  If we are matching $9s^2 + 18s+1 = A(s+1)^2 + B(s+1)$ and we happen to choose $s=0$ to find $A+B=1$ and then choose to match the coefficient of $s^0$, we will again get $A+B=1$, and this new equation is not linearly independent: it gives us no additional information.  So another equation is still needed.

````{prf:example} Repeated Roots, using multiple values of $s$ to find coefficients.

Find $f(t)$ if 

$$\tilde{f}(s) = \frac{3s^3 + 10s^2 + 10s +2}{(s+1)^3(s+2)}$$

We write out the partial fraction expansion

$$
\frac{3s^3 + 10s^2 + 10s +2}{(s+1)^3(s+2)} = \frac{A}{(s+1)^3} + \frac{B}{(s+1)^2} + \frac{C}{s+1} + \frac{D}{(s+2)} 
$$
As always, we multiply by the denominator $q(s)$:

$$
3s^3 + 10s^2 + 10s +2 = A(s+2) + B(s+1)(s+2) + C(s+1)^2(s+2) + D(s+1)^3
$$
By using $s=-1$ and $s=-2$ we can quickly get the coefficients $A$ and $D$:

If we set $s=-1$:

$$
-1 = A
$$

If we set $s=-2$:

$$
-2 = -D \Rightarrow D=2
$$

There are no longer obvious choices for what values of $s$ to use to find $B$ and $C$.  We need two more equations to find the remaining two coefficients.  

We will try other values of $s$.  It is convenient to try $s=0$ (which is equivalent to matching the coefficient of $s^0$).

\begin{align*}
2 &= 2A + 2B + 2C + D\\
  & = -2 + 2B+2C+2 \\
  \Rightarrow B+C&=1
\end{align*}

Another convenient option is $s=1$

\begin{align*}
25  &= 3A+ 6B + 12C + 8D \\
    &= -3 + 6B + 12C +16 \\
\Rightarrow 6B+12C &= 12
\end{align*}

Solving this system gives $B=0$ and $C=1$. Thus

$$
\tilde{f}(s) = \frac{-1}{(s+1)^3} + \frac{1}{s+1} + \frac{2}{s+2}
$$
To take the inverse Laplace Transform for the term with $(s+1)^3$, we will use the $s$-shift.  We start by breaking up the inverse transform into separate pieces:

\begin{align*}
f(t) &= \mathcal{L}^{-1}[\tilde{f}(s)](t)\\
     &= \mathcal{L}^{-1}\left[\frac{-1}{(s+1)^3} + \frac{1}{s+1} + \frac{2}{s+2}\right] (t)\\
     &= \mathcal{L}^{-1}\left[\frac{-1}{(s+1)^3}\right](t) + e^{-t} + 2 e^{-2t}
\end{align*}
Now we apply the $s$-shift:

\begin{align*}
f(t) &= -e^{-t} \mathcal{L}^{-1}\left[\frac{1}{s^3}\right](t)  + e^{-t} + 2 e^{-2t}\\
     &= -e^{-t}\frac{1}{2} \mathcal{L}^{-1}\left[ \frac{2}{s^3}\right](t) + e^{-t} + 2 e^{-2t}\\
     &= \frac{-1}{2} e^{-t} t^2 + e^{-t} + 2 e^{-2t}
\end{align*}

We plot $f(t)$ below.
````


`````matlab
t = linspace(0, 4, 1001);
plot(t, -exp(-t).*t.^2/2 + exp(-t) + 2*exp(-2*t), 'DisplayName', '$f(t)$')
hold on
%plot(t, -exp(-t).*t.^2/2, 'k--', 'DisplayName', '$-\frac{1}{2}e^{-t}t^2$')
%plot(t,  exp(-t) , 'k-.', 'DisplayName', '$e^{-t}$')
%plot(t,  2*exp(-2*t), 'k:', 'DisplayName', '$2e^{-2t}$')

legend('location', 'northeast', 'Interpreter', 'latex')
xlabel('$t$', 'Interpreter', 'latex')
`````

We now consider another example, using the same function, but a different approach to getting the final two coefficients:

````{prf:example} Repeated Roots, matching coefficients of $s$ to solve the partial fractions.

Find $f(t)$ for the same $\tilde{f}(s)$ as before:

$$\tilde{f}(s) = \frac{3s^3 + 10s^2 + 10s +2}{(s+1)^3(s+2)}$$

We have the partial fraction expansion

$$
\frac{3s^3 + 10s^2 + 10s +2}{(s+1)^3(s+2)} = \frac{A}{(s+1)^3} + \frac{B}{(s+1)^2} + \frac{C}{s+1} + \frac{D}{(s+2)} 
$$
Following the same steps we get

$$
3s^3 + 10s^2 + 10s +2 = A(s+2) + B(s+1)(s+2) + C(s+1)^2(s+2) + D(s+1)^3
$$
and by setting $s=-1$ and $s=-2$ as before we have

\begin{align*}
A &= -1\\
D&=2
\end{align*}
We need two more equations to find $B$ and $C$.  

The coefficients of each power of $s$ must be the same on both sides of the equation.  Taking the coefficient of $s^3$, we have $3$ on the left hand side and $C+D$ on the right hand side, and we know $D=2$.  Thus
\begin{align*}
3 &= C+D\\
\Rightarrow C &= 1
\end{align*}
Another equation can come from matching the coefficient of $s^2$:
\begin{align*}
10 &= B + 4C + 3D\\
\Rightarrow B &= 10 - 4-6\\
&= 0
\end{align*}
Alternately, another coefficient that is easy to find is the coefficient of $s^0$, which is equivalent to setting $s=0$.  This would have given $2 = 2A+2B+2C+D$, which would also lead to $B=0$.  This equation would not be useful if we have already used $s=0$ to find another coefficient.

From this point, we can return to the solution method in the previous example to find
\begin{align*}
f(t) &= -e^{-t} \mathcal{L}^{-1}\left[\frac{1}{s^3}\right](t)  + e^{-t} + 2 e^{-2t}\\
     &= -e^{-t}\frac{1}{2} \mathcal{L}^{-1}\left[ \frac{2}{s^3}\right](t) + e^{-t} + 2 e^{-2t}\\
     &= \frac{-1}{2} e^{-t} t^2 + e^{-t} + 2 e^{-2t}
\end{align*}
````


## A quadratic factor that does not factor further.

We will also consider cases where a quadratic term appears that does not factor.  
````{prf:example} A quadratic factor that does not factor further

Find $f(t)$ if

$$
\tilde{f}(s) = \frac{6s^2-30s+40}{(s-1)\left((s-3)^2+2^2\right)}
$$
The quadratic $(s-3)^2+2^2$ does not factor further using real coefficients.  The Partial Fractions expansion we use is:

$$
\tilde{f}(s) = \frac{A}{s-1} + \frac{B(s-3) + 2C}{(s-3)^2 + 2^2}
$$
Multiplying through by the denominator of $\tilde{f}(s)$ we get

$$
6s^2-30s+40 = A\left((s-3)^2+2^2\right) + B(s-3)(s-1) + 2C(s-1)
$$
Choosing $s=1$ gives

$$
16 = A((-2)^2 + 2^2) + 0B + 0C
$$
So $A=2$

Choosing $s=3$ gives

$$
4 = 4A + 4C = 8+4C
$$
and thus $C=-1$.

Finally we need one more equation.  We will take $s=0$:

$$
40 = 13A + 3B -2C = 28 + 3B
$$
So $B=4$.  Putting it together we have

$$
\tilde{f}(s) = \frac{2}{s-1} + \frac{4(s-3) - 1(2)}{(s-3)^2+2^2}
$$

Taking the inverse transform we get

$$
f(t) = 2e^{t} + 4 e^{3t} \cos(2t) - e^{3t} \sin(2t)
$$
which we plot below

````
 

`````matlab
t = linspace(0, 2.5, 1001);
plot(t, 2*exp(t) + 4*exp(3*t).*cos(2*t) - exp(3*t).*sin(2*t), 'DisplayName', '$f(t)$')


legend('location', 'northeast', 'Interpreter', 'latex')
xlabel('$t$', 'Interpreter', 'latex')
`````
