# Extensions to Lagrange Multipliers

## The case $\nabla f = \vec{0}$
In the argument above, we implicitly assumed that given a contour of $f$, the values of $f$ on one side of the contour are larger and on the other side they are smaller.  

It could be that the contours on both sides are smaller or on both sides are larger.  For example, we might have 

$$
f(x,y) = \sqrt{x^2+y^2} - (x^2+y^2)
$$
This increases as the distance from the origin increases until that distance is $\sqrt{x^2+y^2}=1/2$, at which point it begins to decrease again.  This is easier to see in polar coordinates.  If we set $r = \sqrt{x^2+y^2}$ then  $f(x,y) = r-r^2$, which has its maximum at $r=1/2$

Luckily in this case the method still works.  We can show that where $\sqrt{x^2+y^2} = 1/2$ the gradient of $f$ is $\vec{0}$:

\begin{align*}
\nabla f(x,y) &= \frac{\partial \sqrt{x^2+y^2} - (x^2+y^2)}{\partial x} \vec{i} + \frac{ \partial \sqrt{x^2+y^2} - (x^2+y^2)}{ \partial y} \vec{j}\\
&= \left(\frac{1}{2}\frac{2x}{\sqrt{x^2+y^2}} - 2x \right) \vec{i} + \left(\frac{1}{2}\frac{2y}{\sqrt{x^2+y^2}} - 2y \right) \vec{j}\\
\end{align*}

As $\sqrt{x^2+y^2} = 1/2$, this becomes 

\begin{align*}
\nabla f(x,y) &= \left(\frac{x}{1/2} - 2x \right)\vec{i} + \left(\frac{y}{1/2} - 2y \right) \vec{j}\\
&= \vec{0}
\end{align*}

And so we would still find such a point by setting $\nabla f = \lambda \nabla g$, where in this case we would find $\lambda = 0$.


Let's work through this, but this time use the constraint $\frac{x^2}{8} + \frac{63}{8}y^2 -1 =0$.  This constraint is chosen to make some of the calculations below nicer, but they won't be amazingly simple. 

````{prf:example} Find local maxima/minima of $f(x,y) = \sqrt{x^2+y^2}-(x^2+y^2)$ subject to $g(x,y)=x^2/8+\frac{63}{8}y^2-1=0$.
:label: example:Lagrange_ex3

We have $\nabla f = \left(\frac{1}{2}\frac{2x}{\sqrt{x^2+y^2}} - 2x \right)\vec{i} + \left(\frac{1}{2}\frac{2y}{\sqrt{x^2+y^2}} - 2y \right)\vec{j}$.  Simplifying we have:

\begin{align*}
\nabla f(x,y) &=  \left(\frac{x }{\sqrt{x^2+y^2}} -2x\right) \vec{i} + \left(\frac{y}{\sqrt{x^2+y^2}} - 2y \right) \vec{j}\\
\nabla g(x,y) &= \frac{x}{4} \vec{i} + \frac{63}{4}y \vec{j}\\
x^2/8 + \frac{63}{8}y^2-1 &=0
\end{align*}
Setting $\nabla f(x,y) = \lambda \nabla g(x,y)$ yields

\begin{align*}
 \left(\frac{x }{\sqrt{x^2+y^2}} -2x\right) &= \lambda \frac{x}{4} \\
\left(\frac{y}{\sqrt{x^2+y^2}} - 2y \right) &= \lambda \frac{63}{4}y
\end{align*}
Looking at the first equation, we either have $x = 0$ or $ \lambda = \frac{4}{\sqrt{x^2+y^2}} - 8$.  We must consider both possibilities.  
- If $x=0$, then the constraint becomes $\frac{63}{8}y^2-1 = 0$, so $y= \pm 2\sqrt{2}/3\sqrt{7}$.  Thus we get two points $(0 , \pm 2\sqrt{2}/3\sqrt{7})$.  The other points will have nicer values.
- If $\lambda = \frac{4}{\sqrt{x^2+y^2}}-8 = 4 \left(\frac{1}{\sqrt{x^2+y^2}}-2\right)$, then 

$$ y\left(\frac{1}{\sqrt{x^2+y^2}} - 2 \right) = 4\left(\frac{1}{\sqrt{x^2+y^2}}-2\right) 16y $$

From this it follows that $y=0$ or $\frac{1}{\sqrt{x^2+y^2}} - 2=0$.  We must again follow both possibilities.

- If $y=0$, then the constraint becomes $x^2/8 - 1 =0$.  From this we get $x = \pm 2\sqrt{2}$.  This yields the points $(\pm 2 \sqrt{2},0)$.  

- If $\frac{1}{\sqrt{x^2+y^2}} - 2=0$, then we have $\sqrt{x^2+y^2} = 1/2$.  Taking the constraint, we have 

\begin{align*}
x^2/8 + y^2/8 +\left(-\frac{y^2}{8} + \frac{63}{8}y^2\right)-1&=0\\
\frac{x^2+y^2}{8} + \frac{62 y^2}{8} - 1 &=0\\
\frac{1}{32} + \frac{31 y^2}{4} -1 &= 0\\
\frac{31}{4} y^2 &= \frac{31}{32}\\
y^2 &=  \frac{1}{8}\\
y &= \pm \frac{1}{2\sqrt{2}}
\end{align*}
Since $x^2+y^2 = 1/4$ and $y^2 = \frac{1}{8}$, we conclude $x^2 = \frac{1}{8}$, so $x = \pm \frac{1}{2 \sqrt{2}}$. This gives us the four points 

$$
\left(\frac{1}{2\sqrt{2}}, \frac{1}{2\sqrt{2}} \right)
\qquad
\left(\frac{1}{2\sqrt{2}}, -\frac{1}{2\sqrt{2}} \right)
\qquad
\left(-\frac{1}{2\sqrt{2}}, \frac{1}{2\sqrt{2}} \right)
\qquad
\left(-\frac{1}{2\sqrt{2}}, -\frac{1}{2\sqrt{2}} \right)
$$

````

For reference, we plot the constraint in red and the contours of $f$.  The ones corresponding to local minima and maxima are in magenta.  The constraint crosses the contour $f(x,y)=1/2$ without being tangent, but on this contour $\nabla f=\vec{0}$, so we still have $\nabla f = \lambda \nabla g$ where $\lambda = 0$.


`````matlab
clear all
clc
clf
x = [-3:0.1:3];
y = [-3:0.1:3];
[X,Y] = meshgrid(x,y);

G = X.^2/8 + (63/8)*Y.^2 - 1 ;

figure(1)

%the following command plots the points given by the constraint g(x,y)=0
%and it creates a list of the x and y coordinates.
[constraint_coords,c] = contour(X,Y,G, [0,0], 'r');
constraint_x = constraint_coords(1,2:end); %x coordinates of the constraint points
constraint_y = constraint_coords(2,2:end); %y coordinates of the constraint points
axis equal;
xlabel('x')
ylabel('y')
hold on 

F = sqrt(X.^2+Y.^2) -(X.^2+Y.^2);
contour(X,Y,F, 25)
theta = [0:pi/20:2*pi];     %plot the local max/min in magenta.
plot(2*sqrt(2)*cos(theta), 2*sqrt(2)*sin(theta), 'Color', 'm')
plot((2*sqrt(2)/(3*sqrt(7)))*cos(theta), (2*sqrt(2)/(3*sqrt(7)))*sin(theta), 'Color', 'm')
plot(cos(theta)/2, sin(theta)/2, 'Color', 'm')
`````

`````matlab
clear all
clc
clf

x = [-3:0.1:3];
y = [-1:0.1:1];
[X,Y] = meshgrid(x,y);

F = sqrt(X.^2+Y.^2) -(X.^2+Y.^2);

figure(2)
surf(X,Y,F)
hold on


theta = [0:pi/20:2*pi];     %plot the local max/min in magenta.

r = 1/2;
plot3(r*cos(theta), r*sin(theta), (r-r^2)*ones(size(theta))+.00075, 'Color', 'm')
r = 2*sqrt(2)/(3*sqrt(7));
plot3(r*cos(theta), r*sin(theta), (r-r^2)*ones(size(theta))+.00075, 'Color', 'm')
plot3(constraint_x, constraint_y, sqrt(constraint_x.^2+constraint_y.^2) - (constraint_x.^2+constraint_y.^2),'r', 'DisplayName', 'The values of $f(x,y)$ where $g(x,y)=0$') %plot values on the constraint
% legend('interpreter', 'latex')


xlim([-1,1])
ylim([-1,1])
zlim([0,0.5])

daspect([1 1 0.5])

xlabel('$x$', 'Interpreter', 'latex')
ylabel('$y$', 'Interpreter', 'latex')
zlabel('$z$', 'Interpreter', 'latex')

`````
