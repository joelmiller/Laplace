# MAT2VCA Computer Lab 4: stationary points

## Introduction
in this lab we will look at stationary points. These are points where both the $x$ and $y$ derivative of a function are zero.

they can come in maxima,  minima, and saddles.

## Example:

Consider the function $f(x,y) = x^2y + y^3/3 - y -x^2/2$.  


We'll plot the corresponding surface within a distance of radius $2$ from the origin in the $x,y$ plane.

```octave
r= [0:0.05:2];
theta = [0:pi/20:2*pi];

[R, Theta] = meshgrid(r, theta);

X = R.*cos(Theta);
Y=R.*sin(Theta);

%x=[-2:0.1:2];      %if you want to plot over a rectangular region, uncomment these lines.
%y = [-2:0.1:2];
%[X,Y] =meshgrid(x,y);

Z = (X.^2.*Y + Y.^3/3-Y-X.^2/2);
surf(X,Y, Z )

figure(2)

contour(X,Y,Z)
daspect([1,1,1])  %set aspect ratio to 1.
```

There are multiple stationary points in this figure.  Let's try to find them.

\begin{align*}
\frac{\partial f}{\partial x} &= 2xy-x \\
&= 2x(y-1/2)\\
\frac{\partial f}{ \partial y} &= x^2 + y^2 -1
\end{align*}

So the curve(s) with $\partial f/\partial x = 0$ are $y=1/2$ and $x=0$.

The curve(s) with $\partial f /\partial y = 0 $ is $x^2+y^2=1$.

Let's plot these in different colors:




```octave
t = [-2:0.1:2];
x=t;
y = 0.5*ones(size(t));
plot(x,y, 'r');
hold on 

x = zeros(size(t));
y = t;
plot(x,y, 'r')


theta = [0:pi/20:2*pi];
x = 1*cos(theta);
y = 1*sin(theta);

plot(x,y,'b')

daspect([1,1,1])
```

Let's overlay this on top of the contour plots:



```octave
%First the contour plot at much higher resolutions:

r= [0:0.04:2];
theta = [0:pi/20:2*pi];

[R, Theta] = meshgrid(r, theta);

X = R.*cos(Theta);
Y=R.*sin(Theta);
Z = (X.^2.*Y + Y.^3/3-Y-X.^2/2);
contour(X,Y,Z, 50)  %the integer at the end specifies how many contours

hold on 


%now the curves df/dx =0 and df/dy = 0

t = [-2:0.1:2];
x=t;
y = 0.5*ones(size(t));
plot(x,y, 'r');

x = zeros(size(t));
y = t;
plot(x,y, 'r')


theta = [0:pi/20:2*pi];
x = 1*cos(theta);
y = 1*sin(theta);

plot(x,y,'b')

daspect([1,1,1])
```
